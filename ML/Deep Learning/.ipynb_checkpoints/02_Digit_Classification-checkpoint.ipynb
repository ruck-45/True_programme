# AIM -

1. load data from keras dataset
2. build a ANN model lke this (can include hidden layers)

- a Black and White image is just a 2D matrix with values spanning from 0 - 255 for each pixel
- 0.... total Black
- 255 .... total White
- We need to flattern the array first inorder to feed each value to one nuron in Input layer
- Also we need to Scale the values between 0-1 because 0-255 is a huge discrete dataset can cause inaccuracies

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow import keras
```

# Loading Dataset

```python
# MNIST is hand written digits dataset

data = keras.datasets.mnist
dir(data)
```

```python
(x_train , y_train) , (x_test , y_test) = data.load_data()
```

```python
# x_train consists of 60,000 28*28 sized matrix containing data of image
# x_test contains 10,000 images

# its a 3D matrix

print(f'x_train shape ---- {x_train.shape}')
print(f'x_test shape ---- {x_test.shape}')
```

```python
# first image with a 28 rows and 28 columns 2D matrix

x_train[0]
```

```python
# plt.gray()
plt.matshow(x_train[0])

# plt.imshow(x_train[0])
```

```python
y_train[0]
```

# Flattening the Image Arrays

```python
# We need to flattern the array to a 1D aray such that each value can go into one of the nurons.
# from (60000,28,28) [3D] - to - (60000,28*28) [2D] for x_train

x_train_flat = x_train.reshape((len(x_train),-1))
x_test_flat = x_test.reshape((len(x_test),-1))
```

# Scaling

```python
# scaling values to 0-1

train_df = pd.DataFrame(x_train_flat/255)
test_df = pd.DataFrame(x_test_flat/255)
```

# ANN ( with no hidden layer )

```python
# 28*28 Input layers
# 10 output layers

# loss = 'sparse_categorical_crossentropy' 

# because y_train is numerical (1,2,3..), not one hot encoded
# for one hot encoded  - use loss = 'categorical_crossentropy'

model = keras.Sequential([
                            keras.layers.Dense(10 , input_shape = (28*28,) , activation = 'sigmoid') # both o/p and i/p layers
                         ])

model.compile( optimizer = 'adam',
               loss = 'sparse_categorical_crossentropy',
               metrics = ['accuracy'])

model.fit(train_df , y_train , epochs = 5)
```

```python
model.evaluate(test_df,y_test)

# we get 92% accuracy
```

# ANN ( with hidden layers )

```python
# Only scaling

x_train2 = x_train/255
```

```python
#  if we donot want to falttern array sepately we can flatten it inside the model

model2 = keras.Sequential([
                            keras.layers.Flatten(input_shape = (28,28)), # input layer
                            keras.layers.Dense(1000, activation = 'relu'), # Hidden layer
                            keras.layers.Dense(1000, activation = 'relu'), # Hidden layer
                            keras.layers.Dense(10, activation = 'sigmoid') # Output layer
                         ])

model2.compile(optimizer = 'adam',
               loss = 'sparse_categorical_crossentropy',
               metrics  = ['accuracy'])

model2.fit(x_train2,y_train, epochs = 5)
```

```python
model2.evaluate(x_test/255,y_test)
```

# Checking a prediction

```python
plt.matshow(x_test[75])
```

```python
y_test[75]
```

```python
y_predicted = np.array(model2.predict(x_test/255))
```

```python
y_predicted[75]
```

```python
y_predicted[75].argmax()
```

```python
y_predicted = [i.argmax() for i in y_predicted]
```

## Confusion matrix

```python
cm = tf.math.confusion_matrix(y_test,y_predicted)
cm
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(cm,annot = True)
plt.xlabel('y_predicted')
plt.ylabel('y_test')
```
