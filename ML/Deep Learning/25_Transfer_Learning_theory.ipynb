# Transfer Learning

- Using a pretrained model to train for a similar problem
- Example:
    - Here we will use Google's MobileNet model to predict our flower dataset
    - This is possible because Mobile net is an image classification model that can classify over 1000 classes
    - We will use the model but predict for 5 classes of flower

- Mobile Net model


- Usually while building a model from scratch, we have a lot of variables to decide on.
- For example for a image classification problem we have to decide on : 
    - number of convolutional layers
    - number of filters
    - filter size
    - padding and stride
    - Pulling type and size
    - number of dense layers
    - number of nurons in them
    - activation functions
    - dropouts etc...

- For a pretrained model that problem has been solved 90% , it has the perfect parameters to solve similar type of problem
- Like Mobile Net is trained on millions of data and able to classify over 1000 different classes thus, it has almost perfect values regarding to all respect, like:
    - Number of convoluional and deep layers
    - and their parameters and hyperparameters etc...

- We will be only using the hidden layers for transfer learning as the output layer predicts for 1000+ catagories
- we will include our own output layer for  or may add some more hidden layers if we wish

- In transfer learning the hidden layers transfered is FROZEN, means during backtracking , the weights and biasses dont change of the transferred model
- Only the layers we have added ourselfs change their weights and biases

- Frozen hidden layers during transfer learning


- If we wish we can control how much layers to freeze and how much not.
- But usually we freeze the entire model we transfer

- Looking from different angle , the frozen hidden layer just acts as a feature extractor
- They donot change their weights and biases so, for each epoch they act as a giant complex formula giving the feature vector at the end to predict
- The main advantage of transfer learning is that they can generalize properly and give high accuracy in small epochs
