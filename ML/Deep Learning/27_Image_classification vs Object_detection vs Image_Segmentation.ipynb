# Image Classification

- Classifying an entire image into one category


- Here we are classifying the entire image as a dog or not
- Even if there may be present other animals or objects the model will still act on wheather this entire picture is dog or not

# Object Detection

- It is Image Classification along with Localization
- It can detect different objects present in a single image and can locate their positions using a bounding box

# Image Sementation

- This is a more fine grained classification where we are exactly specifying where each pixel is belonging to which class
