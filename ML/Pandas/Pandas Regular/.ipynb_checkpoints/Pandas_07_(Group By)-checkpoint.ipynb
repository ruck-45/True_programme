```python
import pandas as pd
```

# Vizualizing groupby()

```python
df = pd.read_csv("./CSV_files/Csv_weather_by_cities1.csv")
df
```

```python
# doesnt provide us grouped information
# we have to mnually do it for each group

df.temperature[df.city=='new york'].max()
```

```python
# df2 here is a wierd complex oject as mentioned in image below , it has a key in the name of each cities and 
# daraframe corresponding to each key

df2 = df.groupby('city')
df2
```

```python
# its kinda in a dataframe in a tuple form

for i in df2 :
    print(i)
```

```python
for i,j in df2:
    print(i)
    print(j)
```

```python
# We can have groups based on multiple columns 

df3 = df.groupby(['city','event'])
```

## .groups method
- shows groups and the indexes they contain

```python
df2.groups
```

```python
df3.groups
```

## .get_group()
- shows the group selected

```python
df2.get_group('paris')
```

```python
df3.get_group(('paris','Cloudy'))
```
