# CSV Reading

## skiprows attribute
- single number : number of rows to skip
- an array of index : index of rows to skip

```python
#start reading from real header
df_csv = pd.read_csv("./CSV_files/Csv_Revenue2.csv")
df_csv
```

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue2.csv",skiprows = 2)
df_csv
```

```python
# It also removed "Microsoft" and "Apple" entry as they ha index - 5 , 8  resp.

df_csv = pd.read_csv("./CSV_files/Csv_Revenue2.csv",skiprows = [0,1,5,8])
df_csv
```

## header attribute
- lets us choose the header(s) by specifying the row index(es)
- we can have multiple headers
- if there are no header present , we have to specify 'header = none' to invoke default header

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue2.csv")
df_csv
```

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue2.csv",header = 2)
df_csv
```

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue2.csv",header = [2,3])
df_csv
```

```python
# missing header

df_csv = pd.read_csv("./CSV_files/Csv_Revenue3.csv", header = None)
df_csv
```

## prefix attribute
- adds the string provided as prefix to the default header given to dataframe (in case header = None)
- 'header = None' must be specified for it to work

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue3.csv", header = None, prefix='col')
df_csv
```

## names attribute
- lets us give custom column names in case there is a missing header
- 'header = none' must be specified for this to work

```python
#add missing header
df_csv = pd.read_csv("./CSV_files/Csv_Revenue3.csv")
df_csv
```

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue3.csv",header = None,names = ['Company','Product','Sales','Profit %'])
df_csv
```

## nrows attribute
- to read only specific number of rows

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue1.csv",nrows=3)
df_csv
```

## usecols attribute
- to read only specific columns
- the column indexes should be provided in a list

```python
#  Sales had index 2 which did not get read

df_csv = pd.read_csv("./CSV_files/Csv_Revenue1.csv",usecols= [0,1,3])
df_csv
```

## index_cols attribute
- sets the column specified as index
- column can be specified as name / index
- we can also make multiple indexes

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue1.csv",index_col='Company')
df_csv
```

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue1.csv",index_col= [2,3])
df_csv
```

## dtype attribute
- we can change the datatype of specific columns using this
- the column names and their preferred datatype can be given in a dictionary format

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue4.csv", dtype= {'Sales' : float})
df_csv
```

```python
df_csv.Sales.dtype
```

## true_values & false_values
- true_values converts Yes provided into boolean value : True
- false_values converts No provided into boolean value : False

```python
df_csv = pd.read_csv("./CSV_files/Csv_Revenue5.csv",true_values=["Yes"], false_values=["No"])
df_csv
```

## na_values attribute
- lets us specify specific inputs to be read as NAN

```python
#1.Replace all not available terms to None value(NAN)
#2.And replace -1 only in 'sales' to NAN as well
df_csv = pd.read_csv("./CSV_files/Csv_Revenue4.csv")
df_csv
```

```python
#1
df_csv = pd.read_csv("./CSV_files/Csv_Revenue4.csv",na_values = ['not available','n.a.'])
df_csv
```

```python
#2
df_csv = pd.read_csv("./CSV_files/Csv_Revenue4.csv",na_values = {'Product' : 'not available',
                                                 'Sales' : -1,
                                                 'Profit %' : 'n.a.'})
df_csv
```

## keep_default_na attribute
- we can set it to False to stop Pandas changing NaN-recognizable-strings to NaN values

```python
# Reading Normally
# Pandas autometically all NaN-strings to actual Nan values

df = pd.read_csv("./CSV_files/Csv_Revenue6.csv")
df
```

```python
# Now Pandas doesnot autometically parses NaN Values

df = pd.read_csv("./CSV_files/Csv_Revenue6.csv", keep_default_na= False)
df
```

## na_filter attribute ( Sure if Dataset has no NaN values)
- Pandas parses every variable in a dataset in order to identify any posible missing values which it can label as NaN
- It is a time and resource consuming process
- If the dataset we are reading has already been processed and there are no NaN values in it, we can use na_filter to completely bypass the checking round
- This saves great deal of time and computing power if the dataset is considerably large

```python
df = pd.read_csv("./CSV_files/Csv_Revenue6.csv", na_filter=False)
df
```

```python
#converters

def convert_profit(p):
    if p<5 :
        return 0
    return p

df_xl = pd.read_excel("./Excel_files/Excel_Revenue1.xlsx","Sheet1",converters = {'Profit %' : convert_profit})
df_xl
```

# Merge two dataframes to an excel (ExcelWriter)

```python
data1 =  pd.DataFrame({
        'Date' : ['1/1/2020','2/1/2020','3/1/2020','4/1/2020','5/1/2020','6/1/2020'],
        'Temperature' : [51,42,46,53,40,39],
        'Windspeed' : [15,20,21,21,25,24],
        'Event' : ['sunny','sunny','windy','rainy','sunny','rainy']
        })

data2 =  pd.DataFrame({
        'Company' : ['TATA','Amazon','Microsoft','Sony','Disney','Apple','Samsung','Google'],
        'Product' : ['Cars','N/A','Xbox','PlayStation','Movies','iPhones','SmartPhones','Pixel'],
        'Sales' : [26724000,1186325740,555687000,769584400,678563250,512369875,620000000,478326541],
        'Profit %' : [5.752,11.564,4.95,6.154,3.156,4.156,3.456,2.962]
        })

with pd.ExcelWriter("./Excel_files/Excel_Merged_Excel1.xlsx") as writer :
    data1.to_excel(writer,sheet_name='weather_details',index =False)
    data2.to_excel(writer,sheet_name='revenue_details',index =False)
```
