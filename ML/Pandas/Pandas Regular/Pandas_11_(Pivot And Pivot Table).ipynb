```python
import pandas as pd
```

# Pivot helps creating complex dataframes
- index of pivot cannot take duplicates for same column

```python
df = pd.read_csv('./CSV_files/Csv_weather_by_cities1.csv')
df.head()
```

```python
# to change date from str to timestamp format

df.day = pd.to_datetime(df.day)
df.head()
```

```python
# Using Pivot to create complex dataframe

df2 = df.pivot(index = 'day' , columns = 'city')
df2
```

### values attribute
-  pick desired column using values

```python

df2 = df.pivot(index = 'city' , columns = 'day',values = 'windspeed')
df2
```

# Pivot table is used to summerize and aggregate data
- pivot table can take duplicate indexes for same column
- it aggregates the data and applies mentioned operation , (mean by default) for all duplicate indexesand shows under one

```python
df = pd.read_csv('./CSV_files/Csv_weather_by_cities2.csv')
df
```

```python
# by default it provides mean (avg) of all datas under same index

df.pivot_table(index = 'city', columns = 'day')
```

### aggfunc attribute
- to change operation applied in pivot table

```python
# df.pivot_table(index = 'city', columns = 'day',aggfunc = 'sum')
df.pivot_table(index = 'city', columns = 'day',aggfunc = 'median')
# df.pivot_table(index = 'city', columns = 'day',aggfunc = 'count')
```

```python
df.pivot_table(index = 'city', columns = 'day',aggfunc = ['sum','mean'])
```

### margin attribute
- margins argument provides aggregate values of the row/column under All section
- the operation under aggfunc is applied

```python
df.pivot_table(index = 'city', columns = 'day',aggfunc = 'sum', margins = True)
```

### fill_value attribute
- used to fill NaN values in pivot table if any

### dropna attribute
- if it is set : dropna = False, then Pandas will not Drop those rows/ columns that entirely contain NaN values

# Grouper 
- an asset in 'pivot_table' to group things under certain catagory to aggregate data
- there are any functionalities of grouper
- make sure to see in pandas documentation to utilize it more
- the following senario is just a tiny example of use if grouper

```python
df = pd.read_csv('./CSV_files/Csv_weather_data5.csv')
df
```

- above is a csv file having temperature and windspeed of 'New York' for month of 'Jan' and 'Feb'

```python
# converting day from string to timestamp

df.day = pd.to_datetime(df.day)
df
```

```python
#  we want to aggregate data in basis of month (specified by 'freq' argument)

df.pivot_table(index = pd.Grouper(freq = 'M' , key = 'day') , columns = 'city' , aggfunc = 'median')
```
