```python
import pandas as pd
import numpy as np
```

# .loc[ ] method
- used to access rows/columns/both of dataframe using labels/index

```python
df = pd.read_csv("./Csv_files/Csv_Revenue1.csv", index_col='Company')
df
```

```python
df.loc['Google']
```

```python
df.loc[['Google','TATA','Disney']]
```

```python
# First attribute is for rows and second for columns

df.loc[['Google','TATA','Disney'], ['Profit %', 'Sales']]
```

```python
# We can use slicing too

df.loc['Disney':'Google', 'Sales']
```

```python
# These boolean values corresponds to each of the index and returns those index whose value is specified True

df.loc[[False, False, True, True, True, True, False, True]]
```

```python
# We can also use conditions
# returs only products whose profit > 5%

df.loc[df['Profit %']>5,['Product']]
```

# .iloc[ ] method
- used to access rows/columns/both of dataframe using numerical index (indexing created by default in pandas)

```python
df
```

```python
df.iloc[::2,[1,2]]
```

```python
# it can also take boolean inputs

df.loc[[False, False, True, True, True, True, False, True]]
```
