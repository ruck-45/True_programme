```python
import pandas as pd
import numpy as np
```

# append
- concats rows at the end (row wise concatination)

```python
df1 = pd.DataFrame({
    "Name" : ['Tom','Leonard','Sheldon','Brittney'],
    "CGPA" : [4.5,6.14,7.24,5.84]
})

df2 = pd.DataFrame({
    "Name" : ['Sally', 'Bernard', 'George','Yumi'],
    "Section" : ['A','C','B','A'],
    "CGPA" : [4.5,6.14,7.24,5.84]
})
```

```python
display(df1,df2)
```

```python
df1.append(df2)
```

```python
# ignore_index to override merged index

df1.append(df2, ignore_index=True)
```

# join
- concats columns (column wise concatination)

```python
df1 = pd.DataFrame({
    "Name" : ['Tom','Leonard','Sheldon','Brittney','Sally', 'Bernard'],
    "CGPA" : [4.5,6.14,7.24,5.84,8.42,6.48]
})

df2 = pd.DataFrame({
    "Name" : ['Tom','Leonard','Sheldon','Brittney'],
    "Section" : ['A','C','B','A'],
    "age" : [18,17,17,19]
})
```

```python
display(df1,df2)
```

```python
df1.join(df2[["Section",'age']])
```

```python
# how operator to perform : Union(outer / default), intersection(inner), left difference (left), right difference (right)

df1.join(df2[["Section",'age']], how = "inner")
```

```python
# if we have columns with same name, we can make use of lsuffix, rsuffix differenciate their names
# Otherwise it will throw error

# df1.join(df2,lsuffix= "_left")
df1.join(df2,rsuffix= "_right")
```
