# Concept

- While Training ML model, our model can represent one of the three states mentioned below
    - Overfit : Model is trained to reduce Loss function as much as possible, so it takes shape of training dataset.
    - Underfit : Model is too simple and doesn't represent the data at all
    - Balanced Fit : Model accurately representing the nature of the data

# Variance and Bias

#### Variance : 
- It is Based on 'TEST ERROR'.
- It is the measure of how much the 'TEST ERROR' varies of a Model
- 'TEST ERROR' is the error obtained in the Test Dataset

#### Bias : 
- It is Based on 'TRAIN ERROR'.
- It is the measure of how accurately the model captures the data, i.e, the average value of 'TRAIN ERROR' the Model produces
- 'TRAIN ERROR' is the error obtained in the Tain Dataset.

# Venn Diagram Of Bias And Varience

- The White Datapoints describe the prediction of Test Data

- Dimentionality Reduction (PCA)
- L1 & L2 Regularization 
- Ensemble Techniques (Bagging & Boosting)
