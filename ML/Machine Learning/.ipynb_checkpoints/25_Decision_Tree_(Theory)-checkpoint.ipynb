# Decision Tree (Case Study)
- Below we have a dataset of employees with Independent features : Company,job,degree
- The dependent feature is  : wheather their salary is more than 100k or not

- We can form a decision tree taking "COMPANY" feature as the basis, and see the data distribution across all unique companies


- Here facebook has a "PURE SUBSET",i.e., all its datapoints belong to same catagory so no further branching for facebook is required
- However, google and abc pharma still contain mixed data, so they need to be branched out further

- We branched out Google on the basis of the feature : "JOB"
- The "business manager" and the "sales executive" Jobs came out to be "PURE SUBSETS"
- But 'computer programmer" feature still has mixed data, so we branch it further on the basis of "DEGREE"


- We performed similar barnching in the "abc pharma" company
- since we now have pure subset for each branch, we have completed our Decision Tree



# How to select ordering of features?
- In above aproach we could have started the first level of Decision Tree on the basis of "DEGREE" instead Of "Company"
- How to know the ordering at which we choose the features for Decision tree?
- CHOOSE THE FEATURE WHICH LEADS TO LEAST ENTROPY & HIGHEST INFORMATION GAIN



# How Does Decision Tree Work?
- First we take Decision on which feature to consider as basis to split the main database into. (decide the nodes)
- The Decision usally taken by computing some statistical measures like `gini impurity` , `entropy`, `Information Gain` etc...
- Make the tree and repeat the process unless each node has a PURE SUBSET.

# Algorithm Used By Decision Trees

- CART (classification and Regression Trees) : 
    - This algorithm works by splitting node into two child nodes repeatedly.
    - even though the feature may have multiple classes, it will split the node into two catagories.



- ID3 (Iterative Dichotomiser 3)
    - iteratively divides each node into two or more at each steps
    - it only works with descrete or nominal data best

- C4.5 (successor of ID3)
    - unlike ID3 it can work on both catagorical and nominal data

etc...
