# Overview

#### In algebra we used to have equations like -

                                    Y = 2X + 3    , x = independent variable
                                                    y = dependent variable
                                                    
      if we have a set of inputs as - X = [1,2,3,4,5]
      we have output - Y = [5,7,9,11,13]
      
      
#### But here our aim is, given sets of independent variable and dependent variable , we are to determine the "RELATION FUNCTION".

                    if Y = [5,7,9,11,13] & X = [1,2,3,4,5]
                    we have to come up with the relation -
                    
                                Y = 2X + 3   (considering the linear relation is possible)

# Approach

Step 1 : Get the information on Dependent and independent variable.
         Y = [5,7,9,11,13] & X = [1,2,3,4,5]
         
Step 2 : Imagine the relation to be - Y = m*X + b.
         Take random values of m, b at begining (m=0,b=0)
         It will give a line that will completely miss the points at first.
      
Step 3 : Calculate Cost For that perticular set of [m,b], through cost function.
         There are a lot of different types of cost functions :
            - Mean Sqaued error (MSE)
            - Mean Absolute error (MAE), etc...
            
         But here we are taking MSE as it is most widely used.
         
         MSE - Calculate all the errors ( distance of actual points from the predicted line )
             - Square them , add them and find the mean.

###  Image showing Graphical meaning of MSE

###  MSE equation (Cost Function)

Step 4 : We need to reduce this Cost. The less the cost the more accurate the line is to the points.
         To do this we need to find 'GRADIENT' of the 'COST FUNCTION'
         
         Gradient : Derivative of Cost function WRT, its components - m , b we are trying to predict.
                    It is the way Cost changes wrt m , b. 
                    
                    Once we know the slope, we can move the value of m, b in that direction to reduce the cost.
                    This process is known as (GRADIENT DECENT)

###  Gradient calculation

###  Gradient Decent - Graphical sense

#### A slight problem - If we take big fixed steps ,we can miss the minima.

#### Thus we multiply a very small value to decrease the length of steps, its known as "LEARNING RATE"

Step 5 : After calculating the gradient we get the direction at which we need to go to decrease the cost function.
         The "Learning Rate" gives the length of step we are going to take in that direction.
         
         So how do we chamge the m,b ?
         
         m = m - (LR) * (partial derivative WRT m)
         b = b - (LR) * (partial derivative WRT b)
         
         LR = Learning Rate
         
Step 6 : Repeat these operations again and again to desired number of times, until the desired cost minimization is met
