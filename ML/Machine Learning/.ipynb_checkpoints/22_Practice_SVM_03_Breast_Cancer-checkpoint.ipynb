```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

from sklearn.datasets import load_breast_cancer
```

# Dataset

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
breast_cancer = load_breast_cancer()
breast_cancer.keys()
```

```python
df = pd .DataFrame(breast_cancer['data'],columns=breast_cancer['feature_names'])
df['target'] = breast_cancer['target']
df.head()
```

```python
df.shape
```

# Null value imputation

```python
df.isnull().sum().sum()
```

# Normalization

```python
normalizer = MinMaxScaler()
df.update(pd.DataFrame(normalizer.fit_transform(df.drop('target', axis = 1)), columns=df.columns[:-1]))
```

```python
df.head()
```

```python
df.describe()
```

# Train Test Split

```python
x = df.drop('target', axis = 1)
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8)
```

# SVM (kernel = 'rbf') // radial basis function

```python
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(x_train,y_train)
svm_rbf.score(x_test,y_test)
```

# SVM (kernel = 'poly')

```python
svm_rbf = SVC(kernel='poly', degree = 2)
svm_rbf.fit(x_train,y_train)
svm_rbf.score(x_test,y_test)
```

# SVM (kernel = 'linear)

```python
svm_rbf = SVC(kernel='linear')
svm_rbf.fit(x_train,y_train)
svm_rbf.score(x_test,y_test)
```

# SVM (kernel = 'linear)

```python
svm_rbf = SVC(kernel='sigmoid')
svm_rbf.fit(x_train,y_train)
svm_rbf.score(x_test,y_test)
```
