## Aim

Designing a Linear Regression model from scratch that takes two lists, 
X = [.....] // independent variable
Y = [.....] // dependent variable

come up with proper value of m,c

example taken -
X = [1,2,3,4,5]
Y = [5,7,9,11,13]

model should  be able to return value of m ~ 2, b ~ 3

### Single variable Model -

```python
def linear_regression(X,Y) :
    
    m = 0
    b = 0
    new_m = 0
    new_b = 0
    LR = 0.07
    iteration = 10000
    
    for j in range(iteration):
        cost = 0
        gradient_m = 0
        gradient_b = 0
        
        y = [m*i+b for i in X]
        
        for i in range(len(Y)):
            cost += (Y[i]-y[i])**2
            gradient_m += 2*X[i]*(y[i]-Y[i])
            gradient_b += 2*(y[i]-Y[i])
        
        cost /= len(Y)
        gradient_m /= len(Y)
        gradient_b /= len(Y)
        
        print(f"iteration = {j}, Cur_m = {m} ,Cur_b = {b} ,Cost = {cost}")
        
        m -= LR*gradient_m
        b -= LR*gradient_b
        
    print(f"final_m = {m}, final_b = {b}")
        
        
X = [1,2,3,4,5]
Y = [5,7,9,11,13]

linear_regression(X,Y)
```
