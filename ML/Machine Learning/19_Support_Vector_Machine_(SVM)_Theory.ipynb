# Support Vector Machine

- Support Vector Machine is a very popular `Supervised ML Algorithm`
- It can solve both Classification and Regression type problems
- Support Vector Machine draws a hyperplane in n dimentional space such that it maximizes the margin between the classification groups

# Hyperplane

- if the classification has 2 features , the Descision boundary is a  : `line`
- if the classification has 3 features , the Descision boundary is a  : `plane`
- if the classification has n features , the Descision boundary is a  : `Hyperplane` 

#### In general we use the term hyperplane to signify Descision Boundary in all dimensions



# Support Vectors and Margins

- Support vectors : data points that are closer to the hyperplane and influence the position and orientation of the hyperplane.
- Margin [denoted by rho(ρ)] : sum of distances from the hyperplane to the closest data point in either side



# Gamma parameter

- It is the inverse of the radius of influence of a datapoint as a support vector

- High Gamma : only nearby datapoints will influence as support vectors
- low Gamma : faraway datapoints will also influence as support vectors

# Types of SVM
- Linear SVM
- Non Linear SVM

## Linear Support Vector Machine (LSVM)
- If we can separate the classes using a single streight line we use LSVM
- Here we take the hyperplane having maximum margin

- the upper / the right part of hyperplane is considered as positive class
- the lower / the left part of hyperplane is considered as negetive class

- the part of margin belonging to the positive class is represented as  : +D
- the part of margin belonging to the negetive class is represented as  : -D
- Total margin = |2D|

- In LSVM :
    - Equation of the hyper plane : w<sup>T</sup>x + b = 0
    - Equation of the left margin : w<sup>T</sup>x + b = -1    
    - Equation of the right margin : w<sup>T</sup>x + b = +1    

- The distance of each datapoint from the hyperplane is given by : r = ( w<sup>T</sup>x + b ) / ||w||

- The equation of the Linear Support Vector Classifier is :
    - f(x) = sign (w<sup>T</sup>x + b)

### Types of Linear Support Vector Machine ( LSVM )
- Hard Margin Classification
- Soft Margin Classification

#### Hard Margin Classification
- Here the distinction between classes are very clear
- No outliers present
- The LSVM can be easily formed of highest accuracy

#### Soft Margin Classification
- Distinction between the classes is not entirely clear
- Dataset is noisy Outliers present
- LSVM of highest accuracy cannot be formed without the integration of SLACK VARIABLE

- The slack variable in SVM is represented by 'ξ'
- Slack distance is the distance of slack variables (outliers) from their respective margins.
- The SVM line decided after considering the slack vraiables is known as a Soft Margin Classification



## Non Linear SVM
