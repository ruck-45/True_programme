# Decision Tree Regression
- It is an implementation to solve Regression Problems using Decision Tree
- Since we are using sklearn library, The Decision Tree runs on CART Algorithm.
- It Gives a different score each time we execute it, so its not fixed

# Mathematical Formula used By Decision Tree Regression to decide how to split
- In Decision Tree Clasifier we use "Gini Impurity" measure / "Entropy" measure to Split the node to leaf nodes. We constantly try to split such that the measure of entropy / gini impurity are minimum in leaf nodes.
- Here in Decision Tree Clasifier we use "Mean Squared Error" instead. Here we constantly try tominimize MSE error in furthur lef nodes
