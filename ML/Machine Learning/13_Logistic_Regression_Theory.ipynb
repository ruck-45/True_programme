### We ave two type of predictions

## 1. Continuous Prediction
- the predicted value is some number, has a quantity
- linear regression is a continuous prediction
- ex - 
    - pedicting sales price
    - pedicting interview mark
    - predicting stock prices
    - predicting temperature....etc

- these prdiction will give us a number in a range in return , hence Continuous

## 2. Catagorical Prediction
- the prediction belongs to certain class
- Logistic regression belongs to Catagorical Prediction
- ex - 
    - wheather someone will buy an insurance (Y/N)
    - Image classification (cat/dog/car/bus/...)
    - Email is spam or not (Y/N)
