# KNN

- KNN is a Supervised ML algorithm , can perform both classification and resression tasks k number of neighbors.
- KNN calculates the distances of k nearest neighbors surrounding the points and assigns value based on the majority
#### Note : KNN wont work properly if our data has a lot of outliers

# Case Study

### Dataset :

- We have Iris database and we have to classify the flowers into three different catagories : Setosa, Versicolor, Verginica
- On the basis of : Sepal Length & Sepal width




### Visualization

- We plot the dataset on the graph and can see a clear cluster forming aroud each flower type


### Prediction
- Suppose we have The yellow point to predict its flower type
- Based on human intuition, we can clearly see it belongs to type Blue (VerGinica)


### KNN (K Nearest Neighbors

- Here we have implemented KNN algorithm to predict which cluster does the new point belongs

### Taking, K = 3

- When k is set to 3, algorithm considers the 3 nearest neighbors to the point
- Since they belong to same class, it also assigns the point to the same class (Verginica)



### Taking, K = 10

- When k is set to 10, algorithm considers the 10 nearest neighbors to the point
- Here, 7 belong to Verginica, & 3 belong to Versicolor
- The algorithm assigns the majority of the classes to the point (Verginica)



### Taking, K = 20

- When k is set to 20, algorithm considers the 10 nearest neighbors to the point
- Here, 7 belong to Verginica, & 11 belong to Versicolor & 2 belong to Setosa
- The algorithm assigns the majority of the classes to the point (Versicolor)
- But the Correct class is Versicolor.
- This happened because K value exceeded the number of samples in the correct catagory

#### Note : DO NOT take K value TOO LOW or TOO HIGH
#### Note : KNN works fine with multiple dimentions
