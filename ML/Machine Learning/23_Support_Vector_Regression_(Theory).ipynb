# Support Vector Regression
- We can solve regression problems using Support Vector Regression
- It is not commonly used as Linear Reagression is itself a preety solid Regression Alorithm

# Linear Support Vector Regression
- it tries to plot a hyperplane whose margins cover up maximum number of data points
- the positive and the negetive half of the margin are denoted by epsilon(系)
- In Linear Regression we try to minimize the error.
- But here we try to keep error less than the threshold, i.e the margin(系)
- we donot care as long as the error is inside the margin



- the data points through which the margin passes are support vectors

- The equation of the Linear Support Vector Classifier is :
    - f(x) = sign (w<sup>T</sup>x + b)

- In LSVR :
    - Equation of the hyper plane : w<sup>T</sup>x + b = 0
    - Equation of the negetive margin : w<sup>T</sup>x + b = -系   
    - Equation of the positive margin : w<sup>T</sup>x + b = +系   

## epsilon parameter
- it sets the range of the margin to consider


# Non-Linear Support Vector Regression
- we can add non linearity in SVR using Kernel Parameter
- If we choose to use kernel = 'Polynomial', we have to set the degree parameter to specify the degree of the polynomial
- C parameter is used to set Regularization value
