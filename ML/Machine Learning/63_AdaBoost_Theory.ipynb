# Charecteristics Of AdaBoost

## #1 `Stumps` : the week learners in AdaBoost

- While making a Decision Tree we make a full sized tree
- A decision tree takes full advantage of all features a dataset has, to make a full sized tree to make decision

- In Contrast each learner in AdaBoost are trees with just a Node and two leaves
- A tree with just one node and two leaves is called a `Stump`
- Stumps are not great at making accurate classifications
- Unlike Decision Tree which uses all available features to make a decision, A stump uses only one feature to make a decision.
- Thus stumps are technically `weak learners`

## `Step - 1` :  Define Sample Weight

- The first thing we do is to give each sample a weight that indicates how important the datapoint is to be corrrectly classified.
- At start each datapoint has identical sample weight as all are equal at start

- When we will gradually apply boosting, we will change the sample weight accordingly
- The misclassified datapoints will have their sample weight increased as they are more important for the next stump to classify them correctly
- The correctly calssified datapoints will have their sample weight decreased as they are less important for the next stump to classify them correctly, as the previous stump has alredy classified them correctly.

- `Note` : All the Sample Weights Add upto `1` all the time

## `Step - 2` :  Form The Stump

- The Stump ( Which is essentially a decision tree with one node ) will take cosideration of the best feature to split the dataset on basis of Gini-impurity or entropy.

## `Final Step` :  Providing The Final Prediction

- Imagine The Left Stumps Classify the datapoint as `Heart Patient`
- And The Right Stumps Classify the datapoint as `Not Heart Patient`

- We Add their `Amount of Say` for each party
- The party having Highest `Amount of Say` Seals the Final Prediction

- Here Since The `Amount of Say` of Left Stumps are Greater than the Right Stumps, so the datapoint is calssified as `Heart Patient`
