# Naive Bayes

- It is a supervised ML algorithm that is used to predict Classification problems
- It is based on Bayes Theorom
- We call it NAIVE because we make a naive assumption that all features we are considering here to predict the Class are INDEPENDENT.
- Although they may not be independent, it makes our calculation a alot easier and also it works in a generalized sense

# Working Of Naive Bayes Classifier
- y here is the target variable
- X is the collection of all features (x1,x2,x3,....)
- Since we have assumed (x1,x2,x3,....) to be independent, so we can apply the formula : P(A & B) = P(A) . P(B)
- Since the lower half is already fixed from the dataset provided, so it is a constant
- Lastly since 'y' is the target variabale and it has catagories, we are goint to take the catagory having the highest probability among all(argmax)


# Types Of Naive Bayes Classifiers
- Naive bayes Does great on catagorical data
- It doesnot do great on numerical data because it by default assumes the distribution to be gaussian / symmetric

#### 1. Gaussian Naive Bayes : 
- It is used in classification and it assumes that features follow a normal distribution.Because of the assumption of the normal distribution, Gaussian Naive Bayes is used in cases when all our features are continuous.

#### 2. Bernaulli Naive Bayes : 
- It assumes that all our features are binary such that they take only two values. Means 0s can represent word does not occur in the document and 1s as "word occurs in the document" .

#### 3. Multinomial Naive Bayes :
- Its is used when we have discrete data (e.g. movie ratings ranging 1 and 5 as each rating will have certain frequency to represent). In text learning we have the count of each word to predict the class or label

# Example

- suppose we are considering "Titanic Dataset"
