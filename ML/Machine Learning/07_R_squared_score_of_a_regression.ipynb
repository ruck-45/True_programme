# what is residual?
- Residual = Expected value - Predicted value

# What is R - Squared?
- R - Squared is a statistical measure of how close the data are to the fitted regression line.
- It is also known as the coeffcient of determination
- the value is between 0 to 1
- 1 (100%) is the best score, & 0 (0%) is the worst score

- R-Squared Formula



- SS (RES) = Sum Squared Regression Error
- SS (TOT) = Sum Squared Total Error

- y(i) = Expected value
- y(^i) = Predicted value
- y(-) = Mean Expected value

# Note : 
- The Value obtained by calling the .score( ) method of sklearn is actually the R-Squared value of that regression

# Training a Linear Regression Model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('./database/csv_files/Csv_canada_per_capita_income.csv')
df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[['year']],df['per capita income (US$)'])
```

```python
linear = LinearRegression()
linear.fit(x_train,y_train)
```

```python
linear.score(x_test,y_test)
```

# Finding the score using sklearn.matrices.r2_score

```python
y_expected = y_test
y_predicted = linear.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
# It is same as the score predicted by our model

score = r2_score(y_expected,y_predicted)
score
```
