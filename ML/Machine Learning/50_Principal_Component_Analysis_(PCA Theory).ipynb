# Principal Component Analysis (PCA) 

- PCA is an alorithm used to reduce dimentions either to
    - get rid of unnecessery features
    - Or to be able to visualize the dataset with multiple dimentions
- PCA is a rocess of figuring out most important features/ principal components that has the most impact on the target variable.
- PCA doesnt alter the current features, ratheer it generates new features from the current features
- PCA just looks at the data from a different perspective and chooses the axis carefully that each PCA captures the maximum varience in their dimention as possible
- PCA produces the same number of features as the original dataset , but the PCs towards the end capture the least amount of data, so they can be left out
- PCA will not work properly if the data is not scaled/ standardised, because it is highly dependent on the orientation of datapoints in the space

## Example 1
- Suppose we have this dataset where we have to predict the home price
- The price might be dependent on the first four columns : 'town', 'area, 'bathroom', 'plot'
- But the influence of the tree nearby must be very negligible


- We cannot drop that column entirely, because we dont know how much percentage of information it contributes toward the price
- so we use PCA here to extract important informations out of the whole dataset
