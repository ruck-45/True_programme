# Split and Merge

### Split

```python
# Split : we can split the image into individual colors : B,G,R

b,g,r = cv2.split(img)
```

```python
# showing the splitted images

cv2.imshow('image',r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# showing each individual split

plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.title('Blue')
plt.imshow(b)
plt.subplot(1,3,2)
plt.title('Green')
plt.imshow(g)
plt.subplot(1,3,3)
plt.title('Red')
plt.imshow(r)
```

### Merge

```python
# we can merge splitted elements to make images

bgr = cv2.merge((b,g,r))
bgb = cv2.merge((b,g,b))
grg = cv2.merge((g,r,g)) # merge always takes 3 channels or it will produce errors
brb = cv2.merge((b,r,b))
```

```python
# showing the merged images

cv2.imshow('image',brb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Working with ROI (region of interest)
- An image is a numpy array, so if we modify any part of that array, the image gets modified
- If we transpose the array image gets flipped
- So if we have an element we want to copy to another place, we can copy the data of that element and paste it in the region we want it to get copied to

- cordinates of interset found


```python
# here we will copy the circle in the logo and put it in another place
# to do so we must have the location of bothe places, ROI and the place where to put it
# that can be done using cv2 (wont be shown here)

roi = img[156:201,227:273,:]
plt.imshow(roi[:,:,::-1]) # showing the region of interest
```

```python
154-199 + 133
```

```python
# pasting the data of circle in desired place

img[88:133,121:167,:] = roi
plt.imshow(img[:,:,::-1])
```

# cv2.add and cv2.addWeighted
- cv2.add just adds numpy arrays of two images
- cv2.addWeighted adds numpy arrays of two images with weights and biases we have provided (w1\*arr1 + w2\*arr2 + b)
#### Note : both the images must be of the same shape

```python
im1 = cv2.imread('./Images/07.jpg')
im2 = cv2.imread('./Images/08.png')

im1 = cv2.resize(im1,(500,500))
im2 = cv2.resize(im2,(500,500))
```

```python
# image 1
plt.imshow(im1[:,:,::-1])
```

```python
# image 2
plt.imshow(im2[:,:,::-1])
```

### cv2.add( )

```python
added_im = cv2.add(im1,im2)
plt.imshow(added_im[:,:,::-1])
```

### cv2.addWeighted( )

```python
weighted_add_im = cv2.addWeighted(im1,0.9,im2,0.4,-100) # w1 = 0.9, w2 = 0.4, b = -100 (to make darker)
plt.imshow(weighted_add_im[:,:,::-1])
```
