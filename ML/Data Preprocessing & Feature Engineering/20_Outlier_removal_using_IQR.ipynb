```python
import pandas as pd
import numpy as np
```

# Theory

- IQR is the difference b/w the 3rd Percentile ( Q3 ) and the 1st Percentile ( Q1 )

- lower_limit = Q1 - 1.5 * IQR
- upper_limit = Q3 + 1.5 * IQR

# Example

```python
df = pd.read_csv('./Datasets/csv_files/Csv_height_03.txt')
df.head()
```

```python
# statistical summery

df.describe()
```

```python
# calculating IQR

IQR = df.height.quantile(0.75) - df.height.quantile(0.25)
IQR
```

```python
min_thresh = df.height.quantile(0.25) - 1.5*IQR
max_thresh = df.height.quantile(0.75) + 1.5*IQR

min_thresh,max_thresh
```

```python
# Outliers

df[(df.height < min_thresh) | (df.height > max_thresh)]
```

```python
# Clean Dataset

df[(df.height >= min_thresh) & (df.height <= max_thresh)]
```
