# What is Percentile?

- Percentage is the value out of hunderd 

- Percentile is measured taking the highest scorer as 100%
- the percentile score indicates te percentage of population below their score

- As mentioned here Mohan who has scored highest Score has 100% percentile.
- That means 100% students (excluding Mohan) are under Mohan

# Removing Outliers Through Percentile
- Setting a maximum and minimum percentile.
- The datapoints above maximum percentile and below minimum percentile set by us are considered Outliers.

```python
# importing csv

df = pd.read_csv('./Datasets/csv_files/Csv_height.csv')
df
```

```python
# maximum threshold (percentile = 95%, i.e 95% population are below that value)

max_thresh = df.height.quantile(0.95)
max_thresh
```

```python
# minimum threshold (percentile = 0.05%, i.e 5% population are below that value)

min_thresh = df.height.quantile(0.05)
min_thresh
```

```python
# All outliers

df[(df.height>max_thresh) | (df.height<min_thresh)] 
```

```python
# Dataframe without outliers

df[(df.height<=max_thresh) & (df.height>=min_thresh)] 
```
