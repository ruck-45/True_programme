```python
import pandas as pd
import numpy as np
```

# Stack is used to transpose a "Header" to a "Column"

 - a file can have multiple headers 

```python
# pandas by efault reads the first row as header, so the following file came in like this

df = pd.read_excel("./Excel_files/Excel_stocks.xlsx")
df
```

```python
# We can specify our headers by : header = [0,1] , i.e. we have 2 level of header in our file

df = pd.read_excel("./Excel_files/Excel_stocks.xlsx", header = [0,1])
df
```

```python
# To get rid of "Unnamed", we parse the date as index


df = pd.read_excel("./Excel_files/Excel_stocks.xlsx", header = [0,1], index_col = [0])
df
```

```python
df.stack()
```

# Choosing The Header to Transpose using "level attribute" in .stack( )
- By default , .stack( ) transposes innermost header 
- But we can specify the header using "level" attribute.
- "dropna = True" can also be given to drop NA rows. 

```python
df
```

```python
# transposing the Uppermost Header
df.stack(level = 0)
```

### Another Example :

```python
df2 = pd.read_excel("./Excel_files/Excel_stocks_3_levels.xlsx", header = [0,1,2] , index_col = [0])
df2
```

```python
df2.stack() #default
```

```python
df2.stack(level = 0)
```

# Using .stack( ) on single header files will return a series of key:value pair

```python
df3 = pd.DataFrame({"location" : ['bancock','new york', 'shanghai', 'hongkong', 'hawaii'],
                  'population (in million)' : [120,150,75,102,45],
                  'females per 100 males' : [78,94,64,82,165],
                  'weather' : ['cloudy', 'snowy', 'fair', 'humid', 'sunny']})

df3.set_index("location", inplace = True)
df3
```

```python
# Its a series

df3_modified = df3.stack()
df3_modified
```

```python
# converting above series to dataframe

pd.DataFrame(df3_modified)
```

# .unstack( ) undoes what .stack( ) does
