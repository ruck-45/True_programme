# date_range
- It is used to generate a range of date_time series

```python
df = pd.read_csv('./Datasets/Csv/Apple_stock_without_dates.csv')
df.head()
```

-  above document hasno date column, we need to add a date column of datatype Timestamp
-  date ranging the whole "June - 2017" , only business days

## pd.date_range( start , end/periods , freq )
- start : starting date ('YYYY-MM-DD')
- end : ending date ('YYYY-MM-DD')
- peroids : instead of giving the end date, we can specify the number of dates we require and i will give us according to the frequency
- freq : state offset strings for type of timespan

```python
# it skipped 3rd and 4th June because those were business holidays (sat, sun)

Date = pd.date_range(start = '2017-06-01', end = '2017-06-30', freq='B')
Date
```

```python
# using periods to generate the same range as above
pd.date_range('2017-06-01',periods=22,freq='B')
```

```python
# setting the above date_range as index

df = df.set_index(Date)
df.head()
```

# .as_freq( freq, method)
- Allows us to regenerate/convert our dataframe to the frequency we have specified
- it will synthetically produce the results of missing frequencies by the method we have specified
- freq : offset sting
- method :  way to fill up gaps
    - 'backfill'/'bfill'
    - 'pad'/'ffill'

```python
# it adds all missing dates (saturday/sunday) and for value it follows ffill method

df.asfreq('D',method='pad')
```

```python
df.asfreq('W',method='pad') # displays only sunday values because it is the end of the week
```
