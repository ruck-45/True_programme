# Random Forest

- We already know how to make a decision Tree out of a dataset and use it to predict classes
- The Green and Red colours represent two separate classes in the dataset
- We construct a decision tree around that dataset such that the leafnodes are a PURE SUBSEt of one of the catagories.

- Decision Trees are very unpredictable and unrealiable
- Unlike Logistic Regression / SVM, where using identical parameters we get same score each time we execute it
- Decision Tree gives different scores each time we execute it, even passing identical charecters.
- Random Forest also produces same uncertainties but, the majority vote process is more reliable than decision tree

- First we divide datasets into multiple random data subsets


- Then we make Decision Tree for each sub-dataset we created
- Then for Prediction, we pass the input to each tree and take the prediction given by majority of them


## Note :
- We can solve both classification and regression type problems through Random Forest
