# Polynomial Linear Regression
- We call it Polynomual "LINEAR" regression, because it makes use of sklearn.linear_model.LinearRegression( )
- Instead of passing the regular features to the linear model, we instead pass the polynomial version of the features

Example :

- In Multi Linear Regression we pass features : x1, x2, x3, x4, .....
- And the predicted result by the linear model comes out to be : y = C + m1\*x1 + m2\*x2 + m3\*x3 + m4\*x4 ..... 

- In polynomial Regression we pass instead : x1 , x2^2 , x3^3, x4^4, ......
- And the predicted result by the linear model comes out to be : y = C + m1\*x1 + m2\*x2^2 + m3\*x3^3 + m4\*x4^4 ..... 

## Note : 
- Polynomial model can actually be very complex as the dataset increases or the order of the polynomial increases
- The order itself tends to increase the size of the dataset
- Also if not being careful, Polynomial Regressioncan overfit the data (while the linear regression was underfitting)

## Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./database/csv_files/Csv_car_prices2.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test, y_train, y_test = train_test_split(df[['Mileage','Age(yrs)']],df['Sell Price($)']) 
```

```python
plt.figure(figsize=(25,8))

plt.subplot(1,2,1)
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.scatter(df.Mileage,df['Sell Price($)'])

plt.subplot(1,2,2)
plt.xlabel('Age(yrs)')
plt.ylabel('Sell Price($)')
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

linear = LinearRegression()
```

```python
linear.fit(x_train,y_train)
linear.score(x_test,y_test)
```

```python
df2 = pd.concat([df,pd.DataFrame(linear.predict(df.drop('Sell Price($)', axis = 1)),columns=['predicted'])], axis = 1)
df2.head()
```

```python
plt.figure(figsize=(25,8))

plt.subplot(1,2,1)
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.scatter(df2.Mileage,df2['Sell Price($)'], label = 'Expected')
plt.scatter(df2.Mileage,df2['predicted'] , label = 'Predicted')
plt.legend()

plt.subplot(1,2,2)
plt.xlabel('Age(yrs)')
plt.ylabel('Sell Price($)')
plt.scatter(df2['Age(yrs)'],df2['Sell Price($)'], label = 'Expected')
plt.scatter(df2['Age(yrs)'],df2['predicted'], label = 'Predicted')
plt.legend()
```

# Polynomial Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)
```

```python
# it applies the formula of the degree on provided columns
# Exmaples : here order = 2

# Order_2 = (Milage,Age)^0 , (Milage,Age)^1 , (Milage,Age)^2
#         = zero_power , Milage , Age , milage_squared , milage * Age , Age_squared

X = pd.DataFrame(poly_reg.fit_transform(df[['Mileage','Age(yrs)']]),
             columns=['zero_power','Milage','Age(yrs)','milage_squared','milage * Age','Age_squared'])
X.head()
```

```python
x_train, x_test , y_train, y_test = train_test_split(X,df['Sell Price($)'], test_size=0.2)
```

```python
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_train,y_train)
```

```python
# huge improvement in accuracy

poly_linear_reg.score(x_test,y_test)
```

```python
df3 = pd.concat([df,pd.DataFrame(poly_linear_reg.predict(X),columns=['predicted'])], axis = 1)
df3.head()
```

```python
plt.figure(figsize=(25,8))

plt.subplot(1,2,1)
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.scatter(df3.Mileage,df3['Sell Price($)'], label = 'Expected')
plt.scatter(df3.Mileage,df3['predicted'] , label = 'Predicted')
plt.legend()

plt.subplot(1,2,2)
plt.xlabel('Age(yrs)')
plt.ylabel('Sell Price($)')
plt.scatter(df3['Age(yrs)'],df3['Sell Price($)'], label = 'Expected')
plt.scatter(df3['Age(yrs)'],df3['predicted'], label = 'Predicted')
plt.legend()
```
