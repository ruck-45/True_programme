# Theory

- suppose we have following dataset and we applied a simple linear regression on it.
- we can see it underfits the data greatly.



- But in contrast, if we use too complicated model, it overfits the data
- The model is clearly comlicated as its quadratic



- The best model to generalize this dataset is someone which is a little bit complex than a linear regression
- But not too much complicated as a quadratic polynomial regression model



- The only way to make the best fit model achievable is by reducing the weights of heigher order terms



- This is done by the term added at the end of the loss function in L1 and L2 regularization
- The value of this term scales higher with higher theta term ( means, the higher order te model tries to reach, it will punish it by raising the Loss value for it)





- L1 regularization is also known as Lasso Regression
- It uses absolute values to add penalty which is order 1 so... its L1

- L2 regularization is also known as Ridge Regression
- It uses Squared values to add penalty which is order 2 so... its L2

# Note :
-  Prediction we use Lasso and Ridge Regeression to generalize the
-  Lasso and Ridge regression are some of the simple techniques to reduce model complexity and prevent over-fitting

## Lasso Regression : 
- Lasso Regression adds an extra term = [Sum of absolute values all weights] * [constant (lambda)], to the calculation of the cost function

## Ridge Regression : 
- Ridge Regression adds the extra term = [Sum of squared value of all weights] * [Constant (lambda)], to calculation of the cost function

# alpha parameter : 
- this parameter is used to set custom value of the constant (lambda) multiplied to the sum of modified weights in both lasso and ridge regression

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split
```

```python
# loading Dataset

df = pd.read_csv('./database/csv_files/Csv_canada_per_capita_income.csv')
df.head()
```

```python
# no null values

df.isnull().sum().sum()
```

```python
df.shape
```

```python
# scatterplot

plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color = 'r', marker = '+')
```

## Splitting Dataset into training and testing dataset

```python
x_train, x_test, y_train, y_test = train_test_split(df[['year']],df['per capita income (US$)'], test_size=0.2, random_state= 5)
```

```python
x_train.shape , y_train.shape
```

```python
x_test
```

# Linear Regression

```python
linear = LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
# score

linear.score(x_test,y_test)
```

```python
# linear regression plot

plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color = 'r', marker = '+')
plt.plot(df.year,linear.predict(df[['year']]))
```

# Lasso Regression

```python
lasso = Lasso(alpha=1500)
```

```python
lasso.fit(x_train,y_train)
```

```python
# score

lasso.score(x_test,y_test)
```

```python
# lasso regression plot

plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color = 'r', marker = '+')
plt.plot(df.year,lasso.predict(df[['year']]))
```

# Ridge Regression

```python
ridge = Ridge(alpha=545)
```

```python
ridge.fit(x_train,y_train)
```

```python
# score

ridge.score(x_test,y_test)
```

```python
# ridge regression plot

plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color = 'r', marker = '+')
plt.plot(df.year,ridge.predict(df[['year']]))
```

# Observation
- above with a little bit of twicking with alpha value, Ridge and Lasso Regression gave better accuracy than Linear Reagression
- Linear : 80.14 %
- Lasso : 80.25 %
- Ridge : 80.47 %

```python
# plot for all regression

plt.figure(figsize=(25,25))

plt.xlabel('year', fontsize = 30)
plt.ylabel('per capita income (US$)', fontsize = 30)
plt.scatter(df.year, df['per capita income (US$)'], color = 'r', marker = '+')
plt.plot(df.year,linear.predict(df[['year']]), label = 'linear')
plt.plot(df.year,lasso.predict(df[['year']]), label = 'lasso')
plt.plot(df.year,ridge.predict(df[['year']]), label = 'ridge')

plt.legend(fontsize = 30)
```
