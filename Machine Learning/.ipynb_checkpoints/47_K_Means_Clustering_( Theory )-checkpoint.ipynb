# Clustering

- ML Algorithms are classified into Supervised, Unsupervised, Reinforcement Learning.
- Before we have studied Supervised Learning where we were needed to Predict an output WRT many Features Provided
- Clustering is an Unsupervised ML model where we try to find out the underlying structure / cluster in a random dataset provided to us
- We dont have to predict anything here
- We just have to find meaningful groups present within the data

# How does Clustering Happens?
- First we define the number of clusters we want the machine to find, i.e, we set the K value (suppose here K = 2)
- `STEP 1` : Take two random Centroids (since we set K = 2) in the space of the Datset.

- `Step 2` : Map each data-point in the dataset to either of the Centroids.
- This can be done by joining the two centroids and drawing a perpendicular line to that line.
- Any data-points to te left side of the perpendicular is nearer to the left centroid & hence It belongs to left Centroid-Cluster.
- Similarly, Any data-points to te right side of the perpendicular is nearer to the right centroid & hence It belongs to right Centroid-Cluster

- `Step 3` : Now we Adjust the random centroids according to the new cluster formed
- We move the centroids to the actual centers of the newly formed clusters.

- `Step 4` : Again recluster every datapoints based on the new position of the centroids

- `Step 5` : Repeat untill there is no more changes in the clustering happening

# Elbow Method
- This method helps us to find the best possible value of k for best clustering of dataset
- We consider every possoble value of K ( starting from K = 1, where all datapoints are part of a single cluster to... K = n (number of datapoints) where each datapoint has its own cluster.
- The we find SSE = Sum OF Squared Errors for each Cluster WRT their centroids for every K and plot a graph of SSE VS K

- SSE of each clusters and then added for K = 2
- For SSE we add the squared distance of each datapoints from its centroid


- The elbow part of the SSE Vs K graph determines the optimal K value for clustering
