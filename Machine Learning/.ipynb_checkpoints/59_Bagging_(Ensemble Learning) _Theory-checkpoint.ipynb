# Ensemble Learning

#### Real Life Example:
- Sometimes we are unable to take a decision on a cloth we want to buy , so we seek advise from our friends and family.
- We ask many people and take their openion, we take a majority vote out of it and than we take the decision.
- We ask so many different people because, if we take advise from one and only one person, his advise could be potentialy biased.
- This is a typical example of Ensemble Learning.

#### Ensemble Learning In ML
- In ML we use Ensemble Learning to avoid overfitting / high varience condition.
- We train multiple models on subsets of the OG dataset and take a majority vote of the outputs
- In Enseble Learning we use techniques like : `Bagging` & `Boosting`

# Bagging

### The Dataset

- Suppose we have this dataset, training a single model on it can lead the model to overfit
- Overfitting leads to high varience
- The reason high varience is bad, because it makes our model unreliable
- It makes a chance/possibility to get either the lowest loss or the highest
- So there is no true loss of our model, it has a huge range
- The advantage of low varience is that we can define a number which can reflect the true loss value of our model

### Replacement with Resampling Technique

- In order to implement Bagging, we need to divide the whole dataset into smaller datasets such that we can train Models on each one of them and then take the Majority vote of their prediction
- `Replacement with Resampling Technique` is a technique used to divide the original dataset to smaller datasets

- In `Replacement with Resampling Technique` for every sample we choose a datapoint from the dataset giving al datapoints an equal probability to get selected
- Down here we selected 4 samples from OG dataset containing 10 samples, the selection of those 4 samples were irrespective of what has already been chosen



- So in `Replacement with Resampling Technique` a single sub-dataset can have duplicates in them
- Also in `Replacement with Resampling Technique` it can happen that some datapoints could not make it to any of the sub-datasets being created, and remain uninvolved in the training process

#### Bagging In Classification

- We make a certain number of sub-datasets from the OG dataset through `Replacement with Resampling Technique`.
- Then we train a classification-model for each sub-dataset 
- the we take the majority of their prediction, which becomes the final prediction



#### Bagging In Regression

- We make a certain number of sub-datasets from the OG dataset through `Replacement with Resampling Technique`.
- Then we train a regression-model for each sub-dataset 
- the we take the Average of their prediction, which becomes the final prediction



#### Note : Bagging is also known as `Bootstrap Aggregation`
- Bootstraping is the process of dividing the large dataset into smaller datasets
- Aggregation is the process of aggregating the predictions of all individual model to provide the final Prediction

#### Advantage of Bagging
 - The main advantage is that the models trained on the sub-datasets are weak learners
 - They dont have sufficient dataset to overfit features
 - Instead thet tend to Generalize a feature more acuurately

#### oob ( Out Of Bag ) score :
- All those datapoints who failed to appear in at least one of the sub-dataset are collected
- Then they are treated as test data and are scored using the bag_model we have created
- Since these are datapoints the model has never seen before, it acts as a preview of test score

# Random Forest (Bagged Tree) & Bagging

- Random Forest is a bagging technique , but instead of only sampling the rows, it also samples the columns
- Meaning that each sub-dataset has not only has rows selected with `Replacement with Resampling Technique` but also columns.
- Not all subset contain all features/columns
