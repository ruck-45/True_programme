# train data and test data

- Train data : The dataset on which The ML model is trained
- Test data : The dataset on which the accuracy of the ML model is tested

# How Do we generate these train and test datasets from the original dataset?

## Method-1 :
#### train data : The complete dataset
#### test data : Randomly choosen data in the dataset
##### Disadvantage : 
- the test data is already seen by the ML model and the score obtained doesnt reflect how the model will perform on real world datasets.

## Method-2 :
#### train data : Some Part of OG Dataset
#### test data : Remaining Part of OG Dataset
##### Disadvantage : 
- Since we have no control over the way train_test_split splits the dataset, we might run into a situation where some datapoints occuring in test dataset have never occured in train dataset
- As a result ML model might fail to recognize certain classes as they were never present in the train data

## Method-3 : K-Fold Cross Validation
- The Algorithm divides the dataset into K different folds/ segments
- The Ml Model is trained in K-1 folds iteratively and the left out fold is reserved to testing
- In the end the average score is taken and displayed
- K-Fold Cross validation represents the accurate way of how the model will perform in the real world data


# Why To use K-Fold Cross Validation?

- Train Test split can be random
- Sometimes a certain ML algorihm might perform better in a certain train test configuration than the remaining one or vice versa.
- How to know which ML algorithm to use to train the final model?
- K-Fold Cross calidation saves the day by giving us the average value of score considering all folds
- This way we can directly choose the ML model based on the score it provides with K-Fold Cross Validation
