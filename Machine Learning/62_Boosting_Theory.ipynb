# Boosting

- It is one of the `Ensemble Learning` Method used in ML
- Multiple weak learners are trained so that they will generalize on a perticular feature and then their result is cobined to produce the final prediction
- Multiple weak learners combinedly form a strong learner

# Bagging Vs Boosting

### Bagging ( Parallel Ensemble )
- In bagging all weak learners are independent learners
- No learner's prediction affected other learner's prediction
- All learner's Predictions have equal weight/say in final prediction

### Boosting ( Sequential Ensemble )
- In boosting all weak learners are not independent. Some of them are dependent on others.
- A learner's prediction may affected other learner's prediction.
- Different learner's Predictions have different weight/say in final prediction. Some are given more priority than others.

# Working of Boosting

- The Base alogorithm reads data and assigns equal weight to each sample observation
- Each datapoints are given same weightage
- One of the weak learner classifies the data by a `decision stump`
- Boosting by default uses a `Decision Tree` and a `decision stump` is a `Decision Tree` having only 1 branch

- After the classification by the weak learner, the misclassified datapoints are given more weightage
- Then another weak learner tries to clasify the whole dataset again using as `Descision Stump`
- But this time it uses the modified weights to give higher priority to misclassified datapoints

- This process continues for all weak learners
- At last their Predictions are combined together to provide the final Prediction

# Type Of Boosting
