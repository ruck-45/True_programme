```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

# Dataset

```python
pd.set_option('max_columns', None)
pd.set_option('max_rows', None)
```

```python
df = pd.read_csv('./database/csv_files/model_ready_house_price_prediction.csv')
df.head()
```

```python
# Train Test Split

x_train, x_test, y_train, y_test = train_test_split(df.drop('SalePrice', axis = 1), df.SalePrice, train_size=0.8)
```

# Descision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DTR = DecisionTreeRegressor(criterion='mse')
DTR.fit(x_train,y_train)
DTR.score(x_test,y_test)
```

```python
DTR = DecisionTreeRegressor(criterion='mae')
DTR.fit(x_train,y_train)
DTR.score(x_test,y_test)
```

```python
DTR = DecisionTreeRegressor(criterion='friedman_mse')
DTR.fit(x_train,y_train)
DTR.score(x_test,y_test)
```

```python
DTR = DecisionTreeRegressor(criterion='poisson')
DTR.fit(x_train,y_train)
DTR.score(x_test,y_test)
```
