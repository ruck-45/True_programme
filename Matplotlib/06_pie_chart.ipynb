```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
subjects = ['Machine Learning', 'Website Developement', 'Database', 'Ethiacal Hacking', 'Block chain']

np.random.seed(45)
participants = np.random.randint(20,40,len(subjects))

np.random.seed(5)
avg_age = np.random.randint(20,40,len(subjects)) 
```

```python
df = pd.DataFrame({
    'sub' : subjects,
    'participants' : participants,
    'avg_age' : avg_age
})

df
```

# Pie Chart
- it takes only one series/list (numerical) and plots in a pie chart
- labels and other attributes should be given in form of list/series too

```python
plt.pie(df.participants)
plt.show()
```

## Adding labels

```python
plt.pie(df.participants, labels = df['sub'])
plt.show()
```

## labeldistance attribute
- distance of labels from center

```python
plt.pie(df.participants, labels = df['sub'], labeldistance= 1.5)
plt.show()
```

## rotatelabels attribute
- Rotate each label to the angle of the corresponding slice if true

```python
plt.pie(df.participants, labels = df['sub'], rotatelabels=True)
plt.show()
```

## frame attribute
- Plot axes frame with the chart if true

```python
plt.pie(df.participants, labels = df['sub'], radius = 0.5,  frame=True)
plt.show()
```

## center attribute
- the cordinates of the center of the chart
- places the center of the pie chart at that position in axis chart

```python
plt.pie(df.participants, labels = df['sub'],radius= 0.5, frame = True, center= (0.5,0.5))
plt.show()
```

## explode attribute
- allows to separate slices out of pie
- given in for of list for each individual pie element

```python
Explode = [0,0,0.09,0,0.4]

plt.pie(df.participants,labels = df['sub'], explode = Explode)
plt.show()
```

## colors attribute
- to give custom colors
- given in for of list for each individual pie element

```python
color = ['r','k','g','m','y']

plt.pie(df.participants,labels = df['sub'], colors = color)
plt.show()
```

## autopct attribute
- shows the pie coverage in percentage for each item

```python
plt.pie(df.participants,labels = df['sub'], autopct= '%0.2f%%') # 0.2f specidies percentage upto 2 decimal places
plt.show()
```

## pctdistance attribute
- distance of percentage value from center

```python
plt.pie(df.participants,labels = df['sub'], autopct= '%0.2f%%', pctdistance= 0.4)
plt.show()
```

## shadow attribute
- gives shadow to pie chart

```python
explode = [0.1,0,0.2,0,0]

plt.pie(df.participants,labels = df['sub'],explode = explode, shadow = True)
plt.show()
```

## radius attribute
- changes the radius of the pie

```python
plt.pie(df.participants,labels = df['sub'],radius = 1.75)
plt.show()
```

## startangle attribute
- defines the starting angle
- start angle can be given in negetive to show clockwise movement

```python
plt.pie(df.participants,labels = df['sub'], startangle = -90)
plt.show()
```

## textprops attribute
- can add properties (styles) to all texts involved
- takes a dictionary as input

```python
textprops = {'fontsize' : 15,
            'color' : "b"}

plt.pie(df.participants,labels = df['sub'], textprops= textprops)
plt.show()
```

## wedgeprops attribute
- can add properties (styles) to borders / wedges
- takes a dictionary as input

```python
wedgeprops = {'linewidth' : 5,
            'linestyle' : "--",
             'edgecolor' : "r"}

plt.pie(df.participants,labels = df['sub'],radius=1.5, wedgeprops= wedgeprops)
plt.show()
```

## Giving legend

```python
plt.figure(figsize=(10,7))

explode = [0.1,0,0.1,0,0.1]
textprops = {"fontsize" : 15}

plt.pie(df.avg_age, labels = df['sub'], radius= 1.7, shadow=True, 
        explode = explode, autopct = "%0.1f%%", pctdistance = 0.7, textprops= textprops)
plt.legend(loc = 5)
plt.show()
```
