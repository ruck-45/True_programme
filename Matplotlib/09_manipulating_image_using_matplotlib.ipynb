```python
import matplotlib.pyplot as plt
import numpy as np
```

# The "image" module of matplotlib is responsible to read image files

```python
import matplotlib.image as img
```

## Importing an image using imread( )

```python
Image = img.imread("./images/img_01.jpg") # returns the image in array form
Image 
```

```python
Image.shape
```

## Plotting the image using imshow( )

```python
plt.imshow(Image)
plt.show()
```

```python
plt.axis("off") # to turn remove the axis

plt.imshow(Image)
plt.show()
```

## colorbar( )
- adds colorbar to the plot on the colors used in image

```python
plt.figure(figsize=(16,9))

plt.axis("off")
plt.imshow(Image)

plt.colorbar()


plt.show()
```

## cmap attribute
- cmap stands for color map
- it applies various filters to image
- it has various values can be looked up on internet

- cmap cannot take 3D matrix as input,it only accepts 2D

```python
r_image = Image[:,:,0] #only took the "r" part of "rgb" value to convert matrix to 2D
r_image.shape
```

```python
plt.figure(figsize=(16,9))

plt.imshow(r_image, cmap="Accent")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(16,9))

plt.imshow(r_image, cmap="cool")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(16,9))

plt.imshow(r_image, cmap="hot")
plt.axis("off")
plt.show()
```
