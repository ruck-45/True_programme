```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stream = ['Engineering', 'Medical', 'Commerce', 'Humanities', 'Politics']

np.random.seed(61)
number_of_students = np.random.randint(50,100,5)

np.random.seed(45)
average_age = np.random.randint(18,25,5)

np.random.seed(48)
average_salary_in_LPA = np.random.randint(5,20,5)
```

```python
df = pd.DataFrame({
    'stream' : stream,
    'no_of_stud' : number_of_students,
    'avg_age' : average_age,
    'avg_salary' : average_salary_in_LPA
})

df
```

# subplot

- allows us to plot several graphs side by side in a tabular form
- syntax : plt.subplot( no_of_rows, no_of_columns, current_graph_index) // The commas are optional
- write the code of graph after mentioning the intex in subplot above it.

```python
plt.figure(figsize=(16,20))

plt.subplot(3,2,1)
# Line chart

plt.title("stream popularity" , fontsize = 18)
plt.xlabel("carrear options", fontsize = 12)
plt.ylabel("Number of applicants" , fontsize = 12)

plt.plot(df.stream,df.no_of_stud, "o--")

plt.grid()

# ---------------------------------------------------------------------------------------
plt.subplot(3,2,2)
# Histogram

plt.title("average college student age" , fontsize = 18)
plt.xlabel("age", fontsize = 12)
plt.ylabel("frequency" , fontsize = 12)

plt.hist(df.avg_age, rwidth = 0.8, color= 'g')

plt.grid()

# ---------------------------------------------------------------------------------------
plt.subplot(3,2,3)
# Bar chart

plt.title("average age in different streams" , fontsize = 18)
plt.xlabel("stream", fontsize = 12)
plt.ylabel("average age" , fontsize = 12)

plt.bar(df.stream, df.avg_age, width=0.5)

# ---------------------------------------------------------------------------------------
plt.subplot(3,2,4)
# Scatter chart

plt.title("average salary in different streams" , fontsize = 18)
plt.xlabel("stream", fontsize = 12)
plt.ylabel("average salary (in LPA)" , fontsize = 12)

plt.scatter(df.stream, df.avg_salary, c = 'r', s = 55)

plt.grid()

# ---------------------------------------------------------------------------------------
plt.subplot(3,2,5)
# Pie chart

plt.title("average salary in different streams" , fontsize = 18)

explode = [0.2,0.1,0.2,0.1,0.1]
textprops = {
    'fontsize' : 12
}
wedgeprops = {
    'linestyle' : '-.',
    'linewidth' : 10
}

plt.pie(df.avg_salary, labels=df.stream, shadow=True , 
        explode = explode, autopct= "%0.2f%%", textprops=textprops, wedgeprops=wedgeprops)

plt.show()
```
