```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
days = np.arange(1,16)
days
```

```python
np.random.seed(5)

temperature = np.random.randint(25,52,15)
temperature
```

### Plotting Line Graph

```python
plt.plot(days,temperature)  #only this will show the plot in jupyter notebook,
# because of its functionality of showing return values immidiately

plt.show() #it is the actual way to show the graph output
```

### Giving the graph labels and title

```python
plt.title("Month : January")
plt.xlabel("days")
plt.ylabel("temperature")


plt.plot(days,temperature)
```

- Note : we can save the graph image by right clicking on it and using "save image as..." option

### Defining the scales of the graph

```python
plt.axis([0,15,0,50]) #axis([xmin, xmax, ymin, ymax])


plt.title("Month : January")
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(days,temperature)
```

### Adding Style to the graph

#### changing color
- see the codes for different colors in help (shift + tab)

```python
plt.plot(days,temperature, color = "r") #red
```

#### changing marker ( the way points in graphs are marked )
- see the codes for different marker-shapes in help (shift + tab)

```python
plt.plot(days,temperature, marker = "^") #triangle marker
```

#### changing linestyle ( style of line joining all the points in the graph )
- see the codes for different line-style in help (shift + tab)

```python
plt.plot(days,temperature, linestyle = ":") #dotted-line
```

#### changing linewidth ( width of line joining all the points in the graph )

```python
plt.plot(days,temperature, linewidth = 4)
```

#### changing markersize ( size of marker )

```python
plt.plot(days,temperature,marker = "o", markersize = 10)
```

#### Shorthand for : ( color , marker & linestyle )
- order : "color marker linestyle" // without any spaces

```python
# Normal way

plt.plot(days,temperature, color = "y" , marker = "o" , linestyle = "--")
```

```python
# Shorthand

plt.plot(days,temperature, "yo--")
```

### fontsize
- we can change the font size of the title, labels using this

```python
plt.title("Month : January", fontsize = 25)
plt.xlabel("days", fontsize = 20)
plt.ylabel("temperature", fontsize = 13)


plt.plot(days,temperature,"ro--")
```

### legend
- used to give name to a line
- used when there is multiple line-charts in a single graph

```python
plt.plot(days,temperature)
plt.legend(["temperature line"], loc = 4) # loc determines the position of the legend
```

```python
# an alternate way of giving the legend name is to use "label" attribute in plot

plt.plot(days,temperature, label = "temperature line")
plt.legend(loc = 4)
```

### Adding Grids to chart

```python
plt.plot(days,temperature)
plt.grid()
```

### Customizing Grid (color , linestyle, linewidth )
- use this in plt.grid( )

```python
plt.plot(days,temperature)
plt.grid(color='r', linestyle = "--", linewidth = 1)
plt.show()
```

# Multiple line charts

```python
days = np.arange(1,32)

np.random.seed(10)
jan_temp = np.random.randint(40,50,31)

np.random.seed(48)
mar_temp = np.random.randint(25,40,31)
```

```python
plt.title("Mumbai Temperature", fontsize = 15)
plt.xlabel("days",fontsize = 12)
plt.ylabel("temperature",fontsize = 12)

plt.plot(days,jan_temp, "ro--", label = "january")
plt.plot(days,mar_temp, "ko-.", label = "march")
plt.legend(loc = 4)

plt.grid()
```

# Increase Graph size using plt.figure( )

```python
plt.figure(figsize=(16,9)) # takes aspect ratio, always write on top



plt.title("Mumbai Temperature", fontsize = 15)
plt.xlabel("days",fontsize = 12)
plt.ylabel("temperature",fontsize = 12)

plt.plot(days,jan_temp, "ro--", label = "january")
plt.plot(days,mar_temp, "ko-.", label = "march")
plt.legend(loc = 4)

plt.grid()
```
