```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns
```

```python
df = pd.DataFrame({
    'bill' : [12,31,34,45,86,115,234,239,554,589,599,864],
    'tip' : [0,0,2,4,6,9,9,12,45,47,65,78],
    'stayed' : [1,1,1,0,0,1,0,1,0,0,1,1]
})

df
```

# Line Plot
- seaborn takes only Dataframes as input

```python
sns.lineplot(x = 'bill',y = 'tip', data = df)
plt.show()
```

## attributes :

```python
# hue : helps plot graph on multiple catagoris
#  here we have plotted graph beween bill and tip , but catagoried between stayed (0,1 - two catagories)

sns.lineplot(x='bill' ,y='tip' ,data = df, hue='stayed')

plt.title("January Statistics",fontsize= 18)
plt.xlabel("checkout bill", fontsize= 12)
plt.ylabel("tip given",fontsize= 12)
```

```python
# style : will use different style of lines for different catagories

sns.lineplot(x='bill' ,y='tip' ,data = df, hue='stayed', style = 'stayed')
```

```python
# size : will use different width of lines for different catagories

sns.lineplot(x='bill' ,y='tip' ,data = df, hue='stayed', size = 'stayed')
```

```python
# pallette : applies different color maps to plot

sns.lineplot(x='bill' ,y='tip' ,data = df, hue='stayed', palette = 'hot')
```

```python
# dashes : wheather to include dashes in plot or not
# style didnt work here because of dashes

sns.lineplot(x='bill' ,y='tip' ,data = df, hue='stayed',dashes=False, style = 'stayed')
```

```python
# legend : can turn on or off legends

sns.lineplot(x='bill' ,y='tip' ,data = df, hue='stayed',legend= None)
```

```python
# markers : to provide style to markers
# it requires style to be enabled

sns.lineplot(x='bill' ,y='tip' ,data = df,hue='stayed', style = 'stayed', markers= ['o','<'])
plt.show()
```
