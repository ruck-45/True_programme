# Smart Contract Compilation

- .sol file :
    - The solidity program file we have created
- As we compile it, it is divided into two parts:
    - ABI
    - Byte Code
- Both of them follow json format (Key,value) pair

- ABI (Abstract Binary Interface )
    - Using ABI other programs / smart contracts can communicate with our smart contract
    - It has list of all functions which other smart contracts can call / request access to.
    - It acts as a bridge between this contract and others

- Example of ABI of a solidity program

```python
[
	{
		"inputs": [],
		"name": "retrieve", # Function
		"outputs": [
			{
				"internalType": "uint256",
				"name": "",
				"type": "uint256"
			}
		],
		"stateMutability": "view",
		"type": "function"
	},
	{
		"inputs": [
			{
				"internalType": "uint256",
				"name": "num",
				"type": "uint256"
			}
		],
		"name": "store", # Function
		"outputs": [],
		"stateMutability": "nonpayable",
		"type": "function"
	}
]
```

- Byte Code:
    - It is the code that actually gets deployed in the blockchain nad runs in the network
    - It contains all the logic and operations defined inside the smart contract
    - Bytecode consists of all the instructions EVM has to execute 
    - The "object" contains the bytecode (instruction set)
    - "opcode" contains the same information as "object" but not in encrypted format
    - If we decrypt "object" using a bytecode to opcode disassembler we will get contents of "opcode" key

### Important Points

- ABI acts as a bridge between applications and smart contracts
- Byte Code is public and in readable form
- Byte Code is immutable
- We cannot generate Source code using ABI and Bytecode
- Smart contract need not be pulic but, companies make smart contract public to maintain transparency

## Contract Deployment Environments

- There are a lot of environments available in Remix IDE to deploy our smart contracts
- They are : 
    - Javascript Virtual Machine : 
        - Transaction executed in a sandbox
        - Own/Volatile memory blockchain , i.e. if we refresh page , the blockchain resets
        - Ideal for testing
    - Injected Web3 : 
        - Deploy / Run the smartcontract on Ethereum Mainnet or Testnet
    - Web3 Provider : 
        - Connect to a Remote Node and Ethereum Client
