```python
import pandas as pd
import numpy as np
```

#### Note : .sample( ) works as .head( ) but returns random samples from the dataframe

```python
# loading csv file

df = pd.read_csv('./Datasets/csv_files/Csv_bhp.csv')
df.sample(10)
```

- Finding outliers based on price_per_sqft

```python
max_thresh = df.price_per_sqft.quantile(0.99)
max_thresh
```

```python
min_thresh = df.price_per_sqft.quantile(0.01)
min_thresh
```

```python
# Outliers

outliers = df[(df.price_per_sqft>max_thresh) | (df.price_per_sqft<min_thresh)]
outliers.sample(10)
```

```python
# number of outliers

outliers.shape
```

```python
# valid datapoints

df2 = df[(df.price_per_sqft<=max_thresh) & (df.price_per_sqft>=min_thresh)]
df2.sample(10)
```

```python
# number of valid datapoints

df2.shape
```
