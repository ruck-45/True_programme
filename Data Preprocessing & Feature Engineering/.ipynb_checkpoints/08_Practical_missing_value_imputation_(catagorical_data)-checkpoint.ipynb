```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# importing dataset

```python
df = pd.read_csv('./Datasets/house_prices/train.csv')
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df.info()
```

# cleaning the columns with missing value more than 10%

```python
del_cols = df.isnull().mean()*100
del_cols = del_cols[del_cols>10].keys()
```

```python
df.shape
```

```python
df2 = df.drop(del_cols, axis=1)
df2.shape
```

# Catagorical Value Data Imputation ( Mode )

```python
# all columns with object type variables

obj_cols = df2.select_dtypes(include=['O'])
```

```python
# visualizing all object columns

sns.heatmap(obj_cols.isnull())
```

```python
# all object columns having null values

null_obj_cols = obj_cols.isnull().sum()
null_obj_cols = null_obj_cols[null_obj_cols>0].keys()
null_obj_cols
```

## Note : 
- we can use .value_counts( ) to see the number of each unique catagories in a catagorical column

```python
df2['BsmtQual'].value_counts()
```

```python
fill_dict = {i : df2[i].mode()[0] for i in null_obj_cols}
fill_dict
```

```python
# making a copy

df3 = df2.copy()
```

```python
# replacing the NAN values with the mode of each columns

df3.fillna(fill_dict, inplace=True)
```

```python
# heatmap

sns.heatmap(df3.select_dtypes(include=['O']).isnull())
```

# Visualization of changes in data

```python
len(null_obj_cols)
```

```python
plt.figure(figsize=(25,25))

for i in range(11):
    plt.subplot(4,3,i+1)
    plt.hist([df2[null_obj_cols[i]].dropna(),df3[null_obj_cols[i]]], label= ['before', 'after'])
    plt.legend()
```
