## Note : 
- the LabelEncoder and OrdinalEncoder of sklearn label catagorical variables similarly, so either one can be used to label Ordinal Variables

- LabelEncoder
    - It is simple
    - has no parameters to pass
    - takes in 1D array ti fit & transform
- OrdinalEncoder
    - It is same as LabelEncoder, but it has some basic parameters like (catagories, dtype)
    - takes in 1D dataframe (important), it accepts datatframe, but only a single column

# Encoding Ordinal Variables using LabelEncoder
 - label encoder encodes ordinally
 - it first gives priority to numbers (0,1,2,.....)
 - then it gives priority to the alphabetical order

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# making a LabelEncoder object

encoder = LabelEncoder()
```

```python
df2 = df.copy()
```

```python
df2['feedback'] = encoder.fit_transform(df['feedback'])
df2['fav_day'] = encoder.fit_transform(df['fav_day'])
```

```python
df2
```

# Custom Encoding using .map( ) method

```python
df3 = df.copy()
df3['feedback'].unique()
```

```python
df3['fav_day'].unique()
```

```python
feedback_map = {
    'O' : 3,
    'A' : 2,
    'B' : 1,
    'C' : 0
}
```

```python
fav_day_map = {
    'Sunday' : 5,
    'Monday' : 4,
    'Tuesday' : 3,
    'Wednesday' : 2,
    'Friday' : 1,
    'Saturday' : 0
}
```

```python
df3.update(df3['feedback'].map(feedback_map))
df3.update(df3['fav_day'].map(fav_day_map))
```
