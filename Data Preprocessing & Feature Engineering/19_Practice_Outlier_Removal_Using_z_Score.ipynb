```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Datasets/csv_files/Csv_bhp_02.csv')
df.head()
```

# First using Quantile Method

```python
max_thresh = df.price_per_sqft.quantile(0.999)
min_thresh = df.price_per_sqft.quantile(0.001)
```

```python
# Outliers

outliers = df[(df.price_per_sqft > max_thresh) | (df.price_per_sqft < min_thresh)]
print(f'shape : {outliers.shape}')
outliers.sampleple(10)
```

```python
# new dataframe with removed outliers

df2 = df[(df.price_per_sqft <= max_thresh) & (df.price_per_sqft >= min_thresh)]
df2.shape
```

# Using 4-Standard Deviation to find out remaining Outliers

```python
# calculating z score

df3 = df2.copy()

df3['z_scores'] = (df2.price_per_sqft - df2.price_per_sqft.mean())/df2.price_per_sqft.std()
df3.sample(10)
```

```python
# finding the outliers

df3[(df3.z_scores>4) | (df3.z_scores<-4)]
```

```python
# Final cleaning

df4 = df3[(df3.z_scores<=4) & (df3.z_scores>=-4)]
print(f'shape : {df4.shape}')
df4.sample(10)
```
