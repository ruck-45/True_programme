# Using 3 - Standard Deviation to identify outliers
- can be used when the data distribution is Normal/Gaussian (Bell Curve)

### Theory

- It is generally noticed in Gaussian Distribution that :
    - 68% values in a normal distribution reside within 1*SD of the mean
    - 95% values in a normal distribution reside within 2*SD of the mean
    - 95% values in a normal distribution reside within 3*SD of the mean
- So the datapoints lying outside of 3*SD of mean are treated as Outliers and hence are removed

### Z - Score

- Z - Score indicates how many standard deviations away a data point is from the mean.

- Z score = ( datapoint - mean ) / standard deviation
- so after calculating z_score, we can eleminate those datapoints who lie more than 3 std deviation away from mean

# Example

```python
# loading the dataframe

df = pd.read_csv('./Datasets/csv_files/Csv_height_02.csv')
df.sample(10)
```

```python
# frequency distribution of height

# it is observed that the frequency distribution is Gaussian / Normal here

plt.xlabel('Height in inches')
plt.ylabel('Frequency')

plt.hist(df.height, bins=25, histtype='barstacked',rwidth=0.8)
plt.show()
```

```python
# calculating z-score

mean = df.height.mean()
std_deviation = df.height.std()

z_scores = (df.height - mean)/std_deviation

df2 = df.copy()
df2['z_score'] = z_scores

df2.sample(10)
```

```python
# 7 outliers are found

outliers = df2[(df2.z_score > 3) | (df2.z_score < -3)]
outliers
```
