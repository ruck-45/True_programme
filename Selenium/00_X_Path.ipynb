# X Path

- In Selenium automation, if the elements are not found by the general locators like id, class, name, etc. then XPath is used to find an element on the web page
- XPath is an XML path for navigation through the HTML structure
- It is a syntax or language for finding any element on a web page using XML path expression
- XPath can be used for both HTML and XML documents to find the location of any element on a webpage using HTML DOM structure

# Types of XPath

- There are two types of XPath:
    - Absolute XPath
    - Relative XPath

## Absolute XPath
- Direct way of finding the element
- It starts with a single slash ( / ), that means we select element from the root node
- Disadvantage is that it is hardcoded and if any changes made, XPath fails
-  Example :
    - /html/body/div[2]/div[1]/div/h4[1]/b/html[1]/body[1]/div[2]/div[1]/div[1]/h4[1]/b[1]

## Relative XPath
- Relative XPath starts from the middle of HTML DOM structure
- It starts with double forward slash ( // )
- It can search elements anywhere on the webpage, means no need to write a long xpath because we can start from the middle of HTML DOM structure
- Example :
    - //div[@class='featured-box cloumnsize1']//h4[1]//b[1]

# Writing XPaths

- Note : 
    - \* represents all elements
    - We can do indexing like [1],[2],[3],[4].... etc if multiple elements are selected in processs

## Basic XPath
- We can find elements using basic XPath expression : //tag[@attribute = 'value']
- The attribute can be:
    - id
    - class
    - name
    - type
    - value
    - herf
    - src
- Example: 
    - //input[@name= 'uid']

## contains(@attribute,'value')
- syntax : //*[contains ( @attribute,'value' )] # searches for all tags that contain following attribute and value pair
- can be used to find out dynamic elements
- Example : 
    - //*[contains(@class,'menu')]

## text ()
- It is used in place of attribue when we want to find a tag using the text between the tag
- syntax : //*[contains ( text(),'text_between_the _tags' )] # searches for all tags that contain following attribute and value pair
- Example : 
    - //*[contains(text(),'Bank Project')]

## Using or & and
- or & and used to apply multiple constarints regarding attribute-value pair
- or : Union
- and : Intersection
- Example:
    - //a[text()='Python' and @style='color:#343434;']

## starts-with( )
- Finds the elements whose attribute start with given sub-part of the value.
- syntax : tag[starts-with(@attribute,'value')]
- Can be used when the attribute value gets changed on refresh or by other dynamic operations on the webpage.
- Example : 
    - //*[starts-with(@class,'co')]

# XPath Axis Methods
- used to find complex and dynamic elements

## preceding :: 
- Selects the tags comming before the currently selected element
- We can also use indexing here to access individual elements if there are multiple matches
- Example :
    - //*[@type = 'submit']//preceding::input[1]

## following :: 
- Selects the tags comming after the currently selected element
- We can also use indexing here to access individual elements if there are multiple matches
- Example :
    - //a[text ()= 'TensorFlow']//following::a[45]

## ancestor :: 
- The ancestor axis selects all ancestors element (grandparent, parent, etc.) of the current node
- Example : 
     - //a[text ()= 'TensorFlow']//ancestor::div[1]

## descendent :: 
- The descendent axis selects all descendent element (child node , grandchild node, etc.) of the current node
- Example : 
     - //form[@name = 'frmLogin']//descendant::input

## child ::
- The child axis selects all child elements of the current node
- Example:
    - //*[@id="java_technologies"]//child::li

## parent ::
- The parent axis selects the parent of the current node
- Example:
    - //*[contains(text(),'SAP ')]//parent::li

## following-sibling ::
- It finds siblings at the same level of the selected element
- Example:
    - //*[contains(text(),'SAP ')]//parent::li//following-sibling::li

## self ::
- Selects the current node or ‘self’ means it indicates the node itself.
- Example:
    - //*[@type='password']//self::input
