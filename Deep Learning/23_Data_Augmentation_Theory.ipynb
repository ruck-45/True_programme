# Data Augmentation
- Tackles Overfitting

- A Convolutional neural network cannot handle tranformations like : zoom, rotation, contrast etc..
- It specifically requires the dataset to train on to identify them
- Data Augmentation technique generates transformed images with random zoom/rortation/contrast etc... and then uses them in training
- This helps diversify and generalize the model more
- It avoids overfitting and exposes the mdel to similar real life senarios
