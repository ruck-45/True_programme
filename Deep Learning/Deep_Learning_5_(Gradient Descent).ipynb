# ANN through Tensorflow for Insurance Data

```python
import tensorflow as tf
import numpy as np
import pandas as pd

from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Csv_insurance_data.csv')
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size = 0.2)
```

```python
x_train.shape
```

```python
x_train.age = x_train.age/x_train.age.max()
x_test.age = x_test.age/x_test.age.max()
```

```python
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
model = keras.Sequential([
                            keras.layers.Dense(1,input_shape = (2,),activation = 'sigmoid',kernel_initializer='ones', bias_initializer='zeros')
                        ])

model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])

model.fit(x_train,y_train,epochs = 2500)
```

```python
model.evaluate(x_test,y_test)
```

```python
weights , bias = model.get_weights()
```
