# Padding

- We have discussed that during hand written digit recognition we use various filters.
- In case of digit 9 there are three filters in use : the loopy pattern, a vertical streight line, and a diagonal line.


- While applying a filter the size of the image is reduced
- if the image is : (m x n)
- if the filter size is : (a x b)
- then the final image size after filter operation will be : (m-a+1 x n-b+1)


- The biggest problem with filtering from the original image is that the participation from the corner are very little.
- If an important feature is present at the corner, then it will be under represented.


- Thus we use extra layer of padding to make maximum use of the corner pixel.
- Here the original image size is : (5 x 7)
- After adding a layer of padding the final image size becomes : (7 x 9)
- And the filter size is : (3 x 3)
- So after filtering the feature map size becomes : (5 x 7) which is equal to original image size.
- This is known as `Same Convolution`.


- Types of padding : 
    - Valid Convolution: no padding
    - Same Convolution: padding is done in such a way that , the feature map has same size that of the original image
    

# Stride

- Stride is the number of pixel skipped by the filter matrix while filtering from the image.
- The more the stride, the smaller will be the feature map from the image.

- stride = (1,1)




- stride = (2,2)
