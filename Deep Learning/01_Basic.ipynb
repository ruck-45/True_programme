# What is a Neuron? (A Logistic Regression Model)

- We have seen the insurance example, where *'Linear Regression'* failed to predict, so we used *'Logistic Regression'*.
- *'Logistic Regression'* first applied *'Linear Regression'* and Then applied *'Sigmoid function'* on top of that to get desired result.

- A single *nuron* is besically a *Logistic Regression*.
- It has a *linear part* and then it applies a *activation function* on top of that.

- The values we provide to it are called Charecteristics and attributes (here Age)
- It takes the charecteristics / attributes
- Apply linear Regression - ( Y = mX + c )
- Weights ( m ) are applied to scale them according to their importance in contributing to the prediction
- They add bias ( c ) for Flexibility
- Then the whole result is carried over to the activation function (here = Sigmoid) to give a sense of non - linearity to model.
- then the output comes scaled

- In case of Multiple Charecteristics , Multiple weights are taken into consideration , corresponding to their contribution to prediction.
- If a charecteristic does not help determine the prediction, it will have the weight set to Zero ( m = 0 )
- If a charecteristic has a lot of influence over predicting, it will have a high weight and vice versa

# What is a Nural Network ?

 - Its a collection of bunch of Nurons stacked with one another, and also forming Multiple Layers.
 - The Input layer takes the original Charecteristics, Implements Logistic Regression on them then Passess forward
 - All the middle Layers are known as Hidden Layers
 - The last layer is the output layer gives the final result
 - A layer can have multiple Nurons
 - The number of nurons in Output ayer Depends on the number of calsses the output has

## Terminologies:

#### Forward propagation :
- The act of processing the data in one layer and sending the processed data to the next layer to carry on.
- The data to flow from Input Lyer - Hidden Lyer - Output Layer is called Forward Propagation

#### Weights & Biases :
- Weights are the coefficients multiplied with the charecteristics/attributes or the data recived from previous layers
- Weights are used to scale the data to fit the taining output data
- Biases are constants added to make the fitting flexible

#### Activation Function :
- They are applied to linear regression to add non- linearity to the model

#### Loss function :
- loss/cost Function measures the total loss at the end of each epoch of the entire training dataset

#### Gradient Descent :
- Gradient is the measure of how much loss Function is changing WRT numerous Weights & Biases in the Whole Nural Network
- Gradient Descend is the act of figuring out the slope of loss function wrt all Weights and Biases such that we can change them in next epoch so that loss function can reduce and achieve a minima

#### Learning Rate :
- inorder for us to not take large steps and miss the minima in gradient descent, we use a very small number multiplied to reduce our step size, its called the learning rate

#### Backward Propagation :
- The act of finding the gradient of Loss function, then reducing the weights and biases accordingly by travelling back from outpul layer to input layer, so that loss function in next iteration will be less
- this act of learning back from mistake is known as back propagation
