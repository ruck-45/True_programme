# Situation
- We have a model that can calssify dogs and no_dogs
- We have passed a test_dataset inside model to classify images
- The upper row depricts the truth, i.e the original images
- The lower row depricts the predicted value of the model

# Accuracy
- How much datapoints are accurately predicted by the model?
- accuracy = (number of correct predictions) / (total number of predictions)

```python
# here :

correct_predictions = 5
total_predictions = 10

accuracy = correct_predictions/total_predictions
accuracy
```

# Precision
- precision is class dependent , i.e., each class has its own precision
- here we have two classes : dog and not_a_dog
- Precision for a class = (correct predictions of that class / total number of predictions predicted to be that class)

#### When talking about Precision, use predictions as base line

#### Precision of dog class

```python
# total number of dog predictions that were correct : 4
# total number of predictions saying 'dog' : 7

d_precision = 4/7
d_precision
```

#### Precision of not dog class

```python
# total number of not_dog predictions that were correct : 1
# total number of predictions saying 'not_dog' : 3

nd_precision = 1/4
nd_precision
```

# Recall
- Recall just like precision is class dependent , i.e., each class has its own recall
- here we have two classes : dog and not_a_dog
- Recall for a class = (correct predictions of that class / total number of datapoints originally belonging to that class)

#### When talking about Recall, use Original datapoints as base line

#### Recall of dog class

```python
# total number of dog predictions that were correct : 4
# total number of datapoints belonging to class 'dog' : 6

d_recall = 4/6
d_recall
```

#### Recall of not_dog class

```python
# total number of not_dog predictions that were correct : 1
# total number of datapoints belonging to class 'not_dog' : 4

nd_recall = 1/4
nd_recall
```

# F1 Score
- it gives the overall perforane of the model
- it is also class dependent
- It is the harmonic mean of precision and recall
- 2\*(precision \* recall)/(precision + recall)

#### F1 score of dog class

```python
d_f1 = 2*d_precision*d_recall / (d_precision + d_recall)
d_f1
```

#### F1 score of Not dog class

```python
nd_f1 = 2*nd_precision*nd_recall / (nd_precision + nd_recall)
nd_f1
```
