```python
random_state = 456

import pandas as pd
import numpy as np
import tensorflow as tf

from tensorflow import keras
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report
from sklearn. model_selection import train_test_split
```

# Dataset

```python
df = pd.read_csv('./datasets/sonar_dataset.csv',header=None)
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.info()
```

```python
df[60].value_counts()
```

```python
df2 = df.copy()
df2[60] = df[60].map({'M':0,'R':1})
```

```python
X = df2.drop(60, axis = 1)
y = df2[60]

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,stratify=y,random_state=random_state)
```

```python
scaler = MinMaxScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

# Model without dropout

```python
X_train_scaled.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_shape = (60,), activation = 'relu'),
    keras.layers.Dense(30,activation = 'relu'),
    keras.layers.Dense(15,activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])

model.compile(
    optimizer='adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

tb_callback = keras.callbacks.TensorBoard(log_dir = './logs/sonar_minibatch')
```

```python
model.fit(X_train_scaled,y_train,epochs=100,callbacks=tb_callback,batch_size = 10) # Minibatch Gradient Descent approach
```

```python
# clearly model is overfiiting

model.evaluate(X_test_scaled,y_test) # accuracy is worse
```

# Model With Dropout

```python
model2 = keras.Sequential([
    keras.layers.Dense(60,input_shape = (60,), activation = 'relu'),
    keras.layers.Dropout(0.5), # 50% nurons will be dropped
    keras.layers.Dense(30,activation = 'relu'),
    keras.layers.Dropout(0.5), # 50% nurons will be dropped
    keras.layers.Dense(15,activation = 'relu'),
    keras.layers.Dropout(0.5), # 50% nurons will be dropped
    keras.layers.Dense(1,activation = 'sigmoid')
])

model2.compile(
    optimizer='adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

tb_callback2 = keras.callbacks.TensorBoard(log_dir = './logs/sonar_minibatch_dropout')
```

```python
model2.fit(X_train_scaled,y_train,epochs=100,callbacks=tb_callback2,batch_size = 10)
```

```python
# dropout generalizes better

model2.evaluate(X_test_scaled,y_test)
```
