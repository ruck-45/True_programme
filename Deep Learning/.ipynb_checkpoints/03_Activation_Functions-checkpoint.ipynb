# Why activation Function is needed ?

- Here in above figure , sigma charecter represents sigmoid function
- Because the sigmoid function in every level, the data propagated has certain meaning
- Ex :

1. With Activation Function (Sigmoid)

- Suppose, Hidden_layer_1 gives output:
    - s1 = sigmoid(y1) 
    - s1 = sigmoid(m1\*X +b1)
- Then, taking this data Hidden_layer_2 gives: 
    - s2 = sigmoid(y2)
    - s2 = sigmoid(m2\*sigmoid(y1)+b2)
    - s2 = sigmoid(m2\*sigmoid(m1\*X +b1)+b2)

2. Without Activation Function

- Suppose, Hidden_layer_1 gives output
    - s1 = y1
    - s1 = m1\*X + b1
- Then, taking this data Hidden_layer_2 gives
    - s2 = y2
    - s2 = m2\*y1 + b2
    - s2 = m2\*m1\*X + b1 + b2
    - s2 = (m1*m2) * X + (b1+b2)
    - s2 = m3 * X + b3
                                                                       
- We could have taken m3 and b3 directly in the first hidden layer and the hidden layer 2 is not even needed.
- So, if we donot use Activation Function then the meaning of hidden layer is useless.


- The above picture shows the senario if activation functions are not used

## Step function

- Step function takes input and scales them to either 0 or 1
- number >= threshold -----> 1
- number < threshold -----> 0

Here the threshold is set to be Median of the age.

```python
# step function :
# Assuming the number is normalized then the threshold would be 0.5

def step_func(num):
    return 1 if num>=0.5 else 0

print(step_func(0.85))
print(step_func(0.5))
print(step_func(0.47))
```

Disadvantage of Step Function

- 
    - Step function makes the prediction either 0 or 1
    - Thus it only works for either fully activating the nuron or completely deactivating the nuron, nothing in middle
    - This type of behaviour makes it useless for multiclass predictions
    - In the example of predicting hand written digits we may find multiple digits returning 1 for their probability being just above 0.5
    - In such case model cannot predict a single output accurately
    

- 
    - Also the slope of step function is 0 for all inputs
    - This means inputs donot affect gradient function any more
    - Gradients are calculated to update the weights and biases during the backprop process
    - Since the gradient of the function is zero, the weights and biases dont update

## Linear Function



- it is simple linear function
- y = cx

```python
# linear function:
# considering the constant to be 5 here

def linear_fun(num):
    return 5*num

print(linear_fun(10))
print(linear_fun(-2))
print(linear_fun(0))
```

Disadvantage of Linear Function
- Activation funcions are all about introducing non linearity, s linear function doesnt helps in any regard

## Sigmoid Function



- Sigmoid Function takes any number in real number set and scales them between 0 - 1 in a smooth curve
- So it doesnt suffers from disadvantage of step function

```python
# Sigmoid function :
import math

def sigmoid_func(num):
    return 1/(1+math.exp(-num))

print(sigmoid_func(85))
print(sigmoid_func(25))
print(sigmoid_func(2))
print(sigmoid_func(-8))
```

Disadvantage of Sigmoid Function

- The Slope of Sigmoid function in middle is preety significant
- But the slope at the ends are very negligible
- So finding the gradient at such point makes it 0 & renders the whole backpropagation useless
- That slows down learning process infinitely
- This problem is known as VANISHING GRADIENT PROBLEM

## tanh( ) function


- It also takes any real number as Input, but scales them between -1 to 1  in a smooth curve

```python
# tanh( ) function :

def tanh_func(num):
    return (math.exp(num) - math.exp(-num))/(math.exp(num) + math.exp(-num))

print(tanh_func(85))
print(tanh_func(-486))
print(tanh_func(-0.5))
print(tanh_func(8))
```

Disadvantage of tanh( ) Function
- Same as Sigmoid function : VANISHING GRADIENT

## NOTE : 
- It is better to use tanh( ) over sigmoid as tanh( ) centralizes ourdata near zero

## ReLU Function ( Rectified linear Unit )


- Takes any Real number as input and returns the same number if its greater than 0, otherwise it returns 0

```python
# ReLU Function :

def relu_func(num):
    return max(num,0)

print(relu_func(45))
print(relu_func(1.5))
print(relu_func(-0.25))
```

Advantages of ReLU Function
- It is light weight and faster
- That is because it only activates selective neurons , not all
- Also the computation is a lot faster and simpler

Disadvantages of ReLU Function
- Relu function does have partial Vanishing Gradint Problem
- When the value is less that 0 the slope becomes 0 and makes back propagation useless
- This can create dead neurons which never get activated

## Leaky ReLu Function


 - To counter the partial Vanishing Gradient Problem in ReLU , leaky ReLu makes the left half a little tilt

```python
# Leaky ReLU function:
# here the slope is set to 0.1
# When we have option to choose the slope of negetive line , it is known as Parameterised ReLU function

def leaky_relu_func(num):
    return num if num>=0 else num*0.1

print(leaky_relu_func(78))
print(leaky_relu_func(-45))
```

## Exponential Linear Unit (ELU)


- It is also a variant of ReLU
- instead of a linear line for negetive values, ELU uses a log curve
- It returns : 
    - x if x>=0
    - c * (e^x -1) if x<0

```python
# ELU function:
# here the constant is set to 0.1

def elu(num):
    return num if num>=0 else 0.1*(math.exp(num)-1)

print(elu(78))
print(elu(-2))
```

## Swish


- Swish is a lesser known activation function discovered by researchers at Google
- It is as computationally efficient as ReLU and shows better performance than ReLU on deeper models
- Swish Function : 
    - y = x * sigmoid(x) = x/(1- e^(-x))
- The curve of the function is smooth and the function is differentiable at all points.
- This is helpful during the model optimization process and is one of the reasons that swish outperforms ReLU

```python
# Swish function:

def swish(num):
    return num / (1 - math.exp(-num))

print(swish(78))
print(swish(-2))
```

## Softmax Function

- Softmax function is often described as a combination of multiple sigmoids.
- Sigmoid returns values between 0 and 1, which can be treated as probabilities of belonging to a particular class.
- Thus sigmoid is widely used for binary classification problems

- The softmax function can be used for multiclass classification problems.
- This function returns the probability for a datapoint belonging to each individual class.



- Note that the sum of all the output values is 1 in softmax function

```python
# Softmax function:

def Softmax(arr):
    z = [math.exp(i) for i in arr]
    return [i/sum(z) for i in z]

print(Softmax([12,45,-41,6]))
sum(Softmax([12,45,-41,6]))
```

# Choosing the right Activation Function

- Sigmoid functions and their combinations generally work better in the case of classifiers
- Sigmoids and tanh functions are sometimes avoided due to the vanishing gradient problem
- tanh function is better than sigmoid function in some cases

- ReLU function is a general activation function and is widely used
- Leaky ReLU should be used if we encounter dead nurons in case of Relu
- ReLU function should only be used in the hidden layers

- Relu and sigmoid are most popular activation functions used
- usually use relu in hidden layer as its super light and doesnt need much computation
- use sigmoid / softmax at the output layer
