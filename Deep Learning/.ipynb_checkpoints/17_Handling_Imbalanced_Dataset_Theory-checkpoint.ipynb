# Imbalanced Dataset

- Suppose, we are working on a fake transaction detection dataset
- The dataset will obviously be heavily imbalanced, beacause fake transaction is very rare to occure and we are needed to identify it
- Normally training a model will not yeild result because, our goal is not classifying which transactions are ligit, instead we have to identify fake transactions correctly
- Even : ( lambda X : return True ) can provide 95% accuracy as 95% of the datapoints are Ligit
- Where as our model should focus on classifying Fake transactions better

- A normal model trained in imbalanced dataset will have a good accuracy , but that doesnt mean that it classifies all of the classes accurately
- Our goal is to have a good f1 score in each catagory even if overall accuracy is not so good.

# Balancing Imbalanced Datasets

- There are various ways of balancing dataset
    - Undersampling : 
        - reducing sample size of higher class to match the sample size of the lower class
    - Oversampling : 
        - Randomly duplicating datapoints of lower class to match the sample size of higher class
    - Oversampling Using SMOTE : 
        - SMOTE = Synthetic Minority Oversampling Technique
        - SMOTE uses KNN to synthetically generate datapoints for minority class
    - Ensemble Techniques : 
        - Here we divide higher sampled data into batches and train minoriy data individually
        - Then we take a majority vote at the end
    - Focal Loss Techique: 
        - It is a type of loss which gives more weightage to minority class and penalizes majority class
