# Dropout Regularization

- We use Regularization to prevent overfitting of model
- In deep learning we use Dropout Regularization to avoid overfitting
- Here we drop random nurons in each hidden layers for each sample train datapoint
- This way neurons donot get dependent on each other and try to find a more general solution/patern
- Dropout also prevents a prominent feature becoming too dominant
