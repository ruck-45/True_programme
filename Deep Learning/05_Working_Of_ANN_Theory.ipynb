# What do ANN do?

- Suppose we are give with some independent variables (x, here) and a depenedent variable (y, here).
- And we are asked to find a relation between.
- Since above example is an easy linear equation we can deduce the relation easily
- But it is not possible for all situations, because:
    - Not all situations are linear , they may have non linearity
    - Also its not always not pssible that a relation is 100% guarenteed, sometimes we have to force a relation even if there are outliers present
    - Also there can be a lot of independent variables alone which increases complexity
- So its clear that it is humanly impossible to always find a relation between variables if a dataset is given

- Why is the relation needed?
    - Relation is needed because if we get a new set of independent variables then we can predict the value for it using that realtion

#### So the primary task of a ML model is to find the relation between the independent and dependent variables, so that we can predict stuffs.

# Working Of ANN

### Dataset : 
- We need to specify the choice of havin insurance in terms of age and affordability.


### Model : 
- We have used an ANN with a single neuron (i.e., it is a simple logistic regression model)


### Forward Prpagation
- We will set the weights and biases as random numbers at start
- We will pass each pair of Age and Affordability as input and record the predicted value (Y_predicted)
- Then we will record the error value through the error function (Log Loss here)


### End Of Epoch and Total Error Calculation
- Since we are using Batch Gradient Descent ( all datapoints must be considered for error calculation ) , we have to wait for all datapoints to go through above prcess to call it an end of Epoch (Round)
- Then Total Eroor is calculated by adding all errors obtained.
- The final error is the average of Total error : ( Total error ) / ( number of data points involved )


### Backpropagartion
- In last epoch we got certain loss value
- We must prioritize on reducing that loss as it will make our prediction model more accurate
- So , for reducing the loss, we must make changes to the weights and biases such that they prduce more accurate prediction
- we subtract each weight and biases with term  : (learning rate * slope of loss function wrt that weight/ bias)
- learning rate is used to control the updation rate of the weights and biases such that we donot miss the minima.


- This process of learning from the mistakes and updating the weights and biases to make less mistake in next epoch is fundamental of backpropagation.
- We continue above operations unless loss is less enough to declare our model suffeciently accurate at predicting.
