# Image Thresholding
- Thresholding is a popular segmentation technique used to separate an object from its background
- We provide a threshold value of intensity
- it divides the pixels present in the image into two parts:
    - pixels having intensity value lower than the threshold value
    - pixels having intensity value higher than the threshold value
- then we decide the type of thresholding to be done to transform both groups of the image

- cv2.threshold( ) takes parameters : 
    - image
    - threshold value of a pixel
    - max intensity val of a pixel
    - thresholding type
- returns "threshold_value" and "thresholded_image"

### Binary thresholding
- below thresh : overwrite to min
- above thresh : overwrite to max

```python
_,binary_th = cv2.threshold(img1,100,255,cv2.THRESH_BINARY)
plt.imshow(binary_th[:,:,::-1])
```

```python
_,binary_th = cv2.threshold(img2,100,255,cv2.THRESH_BINARY)
plt.imshow(binary_th[:,:,::-1])
```

### Inverse Binary thresholding
- above thresh : overwrite to min
- below thresh : overwrite to max

```python
_,inv_binary_th = cv2.threshold(img1,100,255,cv2.THRESH_BINARY_INV)
plt.imshow(inv_binary_th[:,:,::-1])
```

```python
_,inv_binary_th = cv2.threshold(img2,100,255,cv2.THRESH_BINARY_INV)
plt.imshow(inv_binary_th[:,:,::-1])
```

### Trunc thresholding
- below thresh : unchanged
- above thresh : same as threshold

```python
_,trunc_th = cv2.threshold(img1,100,255,cv2.THRESH_TRUNC)
plt.imshow(trunc_th[:,:,::-1])
```

```python
_,trunc_th = cv2.threshold(img2,100,255,cv2.THRESH_TRUNC)
plt.imshow(trunc_th[:,:,::-1])
```

### Tozero thresholding
- below thresh : makes 0
- above thresh : unchanged

```python
_,tozero_th = cv2.threshold(img1,100,255,cv2.THRESH_TOZERO)
plt.imshow(tozero_th[:,:,::-1])
```

```python
_,tozero_th = cv2.threshold(img2,100,255,cv2.THRESH_TOZERO)
plt.imshow(tozero_th[:,:,::-1])
```

### Tozero thresholding
- below thresh : unchanged
- above thresh : makes 0

```python
_,inv_tozero_th = cv2.threshold(img1,100,255,cv2.THRESH_TOZERO_INV)
plt.imshow(inv_tozero_th[:,:,::-1])
```

```python
_,inv_tozero_th = cv2.threshold(img2,100,255,cv2.THRESH_TOZERO_INV)
plt.imshow(inv_tozero_th[:,:,::-1])
```
