# Nonce

- A bitcoin block consists of : 
    - Block Number : index of the block
    - Data : List of all transactions included in the block
    - Previous Hash : Hash of the previous Block
    <br> <br>
    - Nonce : It is the special number that the miners have to find out which solve sthe problem given to them

- Usually the problem consists of starting the hash of the given block with certain number of zeros
- The SHA256 Algorithm takes all above mentioned containts of a block and calculates the hash of the current block
- If the Hash satisfies the requirement, then the block gets successfully added to the chain

- Nonce is actually the proof of work miner shows which prooves that the miner has mined that block

# Target

- Target is a number that the final hash of the block is compared with
- It is a number from which the block hash must below for the block to be added to the blockchain
- The target adjusts itself every 2016 blocks (roughly 2 weeks) to ensure the blocks are mined on an average every

# CPUs, GPUs, And ASICs

- Mining has been competative since start
- But now it has grown much more aggressive because of advance in technology
- The better technology someone has the better is his/her chance to get profit out of mining

### CPU
- Before CPUs were used for mining 
- Thy can check <= 1 Million hash per sec
- But as per todays standard thats very low and ineffective


### GPU
- GPUs have significantly higher processing power than CPUs, for thsi reason most of the miners have jumped ships
- GPUs processing power is in the range of <= 1 billion hashes per second


### ASICs ( Application Specific Integrated Circuits )
- These are circuits specifically designed for mining
- They can mine at a processing speed of <= 1000 Trillion hashes per second
- But they are crazy expensive


# Mining Pool

- Suppose there are regular group of miners, using GPUs for mining
- But one day they have competitors with huge budget and using ASICs
- This will make the mining environment one sided and unfair


- So small miners make pools whrere they mine blocks collectively increasing their processing power
