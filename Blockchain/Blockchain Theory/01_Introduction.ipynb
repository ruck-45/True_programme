        - automobiles replacing carts

- Just like Internet revolutionized communication , Blockchain revolutionized trust

- Blockchain is a
    - distributed : it is decentralized (everyone has their own copy of entire blockchain)
    - immutable : cannot be changed / infringed
    - transperent : anyone can view all transactions
    - ledger : a transaction record

# Application of Blockchain

1. Product Tracking
    - Suppose we are buying a product that represents a brand
    - But we donot trust the seller
    - How to varify if the product is legit?
    - This is where product tracking through blockchain comes into play

    - In denmark its already in effect
    - There in every product packet exists a barcode that represents the blockchain of that product
    - It provides every possible information about the product, like its quality tasted or not, and places it has been through etc

2. Smart Contracts
    - Suppose there are two parties under a contract
    - But the trust between is not so solid
    - There might be a situation where one of them refuses to keep their part of the bargain
    - here Samrt contracts come into play
    - They are transperent and deterministic
    - That means if one holds their end of their bargain then smart contract will immidiately initiate the other end of the bargain
    - Smart contracts are immutable , so they cant be changed
    - They remove the necessety of trust in a deal

3. Healthcare Systems
    - In healthcare Industry blockchain can help a great deal
    - Suppose a man goes to Hospital A beacause it is near to his home
    - All his dignosed data is stored in database of hospital A
    - Suppose he suddenly fell sick in another city and is in serious condition
    - If Hospital  A could have maintained a blockchain for storing patient records then Hospital B could have easily accessed it and started dignosing the patient emmidiately without redoing bunch of tests.

3. Transaction Of money
    - In a cenralized system like banking system, if we want to transfer money from one country to another, we need to go through several procedures
    - Ther are also several mediator banks in between who will charge fee
    - Blockchain removes all these complications and makes money transfer seamless
