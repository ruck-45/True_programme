# Technology , Protocol / Coin & Tokens

- Technology : 
    - Here The technology used is Blockchain
    - Blockchain is a decentralized immutable ledger technology maintained by miners
    


- Protocol : 
    - The system developed over the proposed technology are called Protocol
    - They are called protocol because all of them are based on the same technology but their behaviour differs because they have different rules to follow or protocols to follow
    - Some example of protocols : 
        - Bitcoin
        - Ethereum
        - Waves
        - Polkadot etc..
        


- Coins : 
    - Each Protocol has a coin of its own to carry out transactions
    - A protocol can expand from being just a coinbase (like Bitcoin) to a whole system (like Ethereum)

- Tokens :
    - Tokens are blockchain protocols who donot have a protocols of their own
    - Means they have not built their own blockchain model like : Ethereum, Bitcoin, Polkadot etc..
    - Instead they act as an asset representation for other protocols
    <br> <br>
    - For Instance : 
        - If automobile is a Technology
        - Then cars are protocols
        - Owening a car , i.e, owening a coin of the protocol requires a lot of maintainance to be done by us
        - We need to check for gas, damages, servicing and repairments every now and then
        - But instead we `RENT` a car and suddenly we donot need to maintain the car, the maintainance is taken care of
        - The rented car is not our car , but it is a representation of ownership of a temporary asset
    <br> <br>
    - Example : 
        - Tron
        - Tether
        - Chainlink
