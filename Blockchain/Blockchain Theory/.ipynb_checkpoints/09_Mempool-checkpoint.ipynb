# Mempool

- Mempool is a list of all un-registered transactions
- Un-registered Transactions are transactions, already performed but are not added to the block chain
- It is the responsibility of Miner to include the transactions in The Mempool in the next block

- Mempool has millions of transactions in it as they are happeneing simultaneously
