# Monetary Policy

- Monetary Polcies are decided by the central authority regarding the flow of money and production of new currency
<br><br>
- Similarly Satoshi Nakamot has defined some Monetary Policy for Bitcoins
- Since Bitcoin is a decentralized network, so there is no central governing body maintaining these policies
- They are self executed from the protocol written under bitcoin

 # Bitcoin Monetary Policy

## The Halving

- The Block Reward for the Bitcoin will half itself, for every 210k block added
- It happens once in every 4 years
- It is decided to limit total number of Bitcoin reserves in the world


- There is no other way of creation of new bitcoin coins except for the block rewards which are generated from thin air and are added to the bitcoin economy
- To prevent infinite expansion and maintain some value to bitcoin it has been limited by this policy.
- It has been predicted that till 2140 the block reward will plummet to 0


- So the Supply cap of Bitcoin is set to 21 Million
- So theoratically there will never be more than 21 million bitcoins in the world

#### So does that mean miners will not get rewards anymore? will that be death of bitcoin?
- No because the main income of miners come from the transaction fee included with each transaction.
- So miners will still get paid even if block rewards cease to exist

## Block Frequency

 - This policy ensure that it will take on an average of `10 minutes` to create (mine) a new block
 - This is set so because it affects the halving policy
 - So the block difficulty dynamically changes to keep up with this policy
