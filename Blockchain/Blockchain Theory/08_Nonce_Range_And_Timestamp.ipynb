# Nonce Range

- There is a Problem Here.
- Nonce is a 32 bit number
- So total number of Nonce availale to a miner to check for hases is 2<sup>32</sup> ~ 4 billion



- But A SHA256 hash string is a 64 digit long string each representing a hexadecimal number
- So Total number of possible Hash strings are : 16<sup>64</sup> which is >>>>>>>>>>>> 4 billion attempts we have with nonce


- This creates a problem that we donot have enough nonce to check for every possible Hash string
- Suppose a Miner has processing power of  : 10<sup>8</sup> hashes per sec (which is normal for a CPU) then he will exhaust the whole Nonce set in 40 seconds
- Also the main problem is that what if our solution doesnt lie in the range of the Nonce?
- What if a higher Nonce is require to solve the problem, will the mining of the block stop entirely?


- Here Timestamp comes into play

# Timestamp

- Problem #1 : a miner is able to exhaust an entire nonce range using a simple cpu in 40 sec
- Problem #2 : range cannot entirely cover the range of hash combination

- To tackle this we have introduced another field in the block of the blockchain : `TimeStamp`

- Timestamp usually contains Unix time
- Unix time is a system for describing a point in time.
- It is the number of seconds that have elapsed since the Unix epoch, excluding leap seconds.

- Solving Problem 1

    - Suppose the computation speed of a miner's system is 10<sup>8</sup> per second
    - Then at the end of a second it will complete upto : 10<sup>8</sup> hashes
    - Our total hashes Nonce can produce is : 2<sup>32</sup> ~ 4 billion ~ 10<sup>9</sup>
    - So in a second we have exhausetd only 1/10 th of Nonce available to us
    - But at the end of each second , the timestamp changes creating an avalanche effect which replenishes the number of Nonces available to us
    - In this manner Nonce never really gets exhausted

- Solving Problem 2
    - With each changing Timestamp the Nonce gets replenished and the Nonce points to a different set of hash strings
    - So it is eived that if enough time has passed, we can map every single sequence in hash string with the help of Nonce and Timestamp

- New Problem : 
    - Mining with CPU has been tackled, but current minin speed is near 240 Million Trillion Hashes oer second
    - It is nearly around : 10<sup>20</sup> hashes per second
    - So 4 billion Nonce will be exhaused in ~ 10<sup>-10</sup> seconds <<<<<<< 1 seconds
    - And the nonce wil take the rest time to replenish once again
    - So what should miners do in that idle time? should they wait?

- This problem is solved using MemPool
