# Blockchain Mining

- When someone makes a transaction in blockchain, the transaction is saved in a `Mempool`
- Miners collect certain transactions from the mempool and create a block
- Then they go on a timed race to solve a mathematical problem what produces the proof of work for that block
- The first miner that solves the block adds it to its chain and then broadcasts it to its peers
- Then other miners validate the block and then tehy add it to their chain
- If the block gets varified by everyone and gets updated everywhere then it officially the next block of the blockchain
- When the block gets added the miner receives a small reward in form of the blockchain currency

# Byzantine General's Problem

- Ther is a problem that surfaces in peer to peer network
- It is called Byzantine general's Problem
- Since in decentralized network there is no server to confirm the validiy of the message, we need to depend on our peers for valid information

- In Byzantine general's case there is a commander who is responsible to announce the next course of action
- All the other lieutenants are posted across the kingdom and they are supposed to carry out the command given by the commander
- There is no direct way of communication so they are dependent on messengers
- Messangers from commander to all the lieutenants takes different times
- lieutenants can also send messages to one another
- All lieutenants are supposed to convey the message received to each other

- Suppose the commander announced "ATTACK" means he expects a coordinated attack from all sides
- But all lieutenants are not faithfull, some of them are traitors
- Suppose a traitor got message first and brodcasted "RETREAT" message to all the lieutenants
- Now other lieutenants will get both the message "ATTACK" and "RETREAT", now which order to follow?

- Above dillema is the heart of P2P network
- Blockchain nodes are continuously bombarded with various candidates to be the next block
- How to confirm which one is the correct next block and which ones are fake blocks made by hackers?

- The above problem is solved by consensus/ majority vote
- The block whose broadcast is done by multiple peers, i.e, majority of the peers accept it as the next block then that perticular block is the ligit next block in the blockchain

- Also in case of Byzantine General's case they were able to perform coordinated attack because they relied on majority voted chain of command
- The traitors were less in number so their fake broadcast got outnumbered

# Consensus Protocol : Proof Of Work (POW)

- In Mempool there must be a large number of transactions waiting to be added to he main blockchain
- Since miners are responsible of adding new blocks to the blockchain, they must decide which transactions should be added to the next block
- But this creates a problem, if the process of adding block will be too easy, it will resulting in multiple candidates for next block
- To remove this complicacy blockchain makes miners solve a computationally exaustive mathematical problem , and whoever gets the answer early gets to decide the next block of the blockchain

- The mathemetical problem generally involves finding "Proof Of Work", which is a unique number for which the hash of that perticular number starts with a fixed number of zeros
- Ofcourse the hash of the block depends on : 
    - The block number
    - Previous block hash
    - Message (transactions)
    - Proof Of Work
- The more number of zeros required , the more difficult is the problem, as the number of possibilities increarse significantly
- As a result it requires more effort and time to find the solution : "Proof Of work"

- The number of starting zeros are decided dynamically
- If number of active miners are less then to make the problem simpler it demands less number of zeros
- If number of active miners are more then it makes the problem harder by increasing the number of starting zeros
- It fluctuates the difficulty such that each new block gets produced in a fixed span of time (say bitcoin adjusts its difficulty such that each new block gets added in every 15 minutes)

- In addition to the dynamic difficulty the transactions choosen by the miner also plays the important role
- For some lucky combination of transactions, he/she might get proof of work sooner that anybody else
- For miners it is like a miniature lottery
- In a fixed interval of time they get a chance to solve a problem and whoever gets the solution faster wins

- The process of Mining is a very power intensive task
- Not only it consumes time, it also uses insane amount of computation and electricity
- So for all the efforts the miner gets a reward for successfully mining a block
- Along with that the miners get transaction fees included in the transactions he/she decided to put in his/her blockchain

# Consensus Protocol : Longest Chain rule

- Sometimes there can be a case such that two miners successfully mined the next block chain
- Both blocks are different
- As we know that as soon as the mine a block they add it to their chain and broadcast

- Since some peers are near an dsome are far away, there is a chance some are updated with either version of the blockchain
- This creates a conflict, but since both of them are valid blocks nothing can be done


- So this kinds of conflicts are solved by the longest chain rule
- Longest chain rule states that if in the same network there exist different versions of the blockchain, then the version with the longest chain is regarded as the main chain
- Here the consensus agrees the chain having the most work put into it as the main chain
- If the chain lengths are equal, it will resolve it when the next block comes in either of tgis blocks


- Discarded blocks are known as orphan blocks
- The miner who has mined the network will not get any reward
- For this rule only it is advised to wit for 6 confirmations (6 more blocks to get added) before considering payment to be successful
- Otherwise the problem of `double payment` can arise
