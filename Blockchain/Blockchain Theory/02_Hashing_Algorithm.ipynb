# Hashing Algorithm

- A hashing algorithm is a function that takes arbitary amount of input and provides a fined length output
- Blockchain used a Cryptographic Hashing alorithm for its security

- Charecteristics of a Cryptographic Hashing Function:
    - One way :
        - We can get the hash-string from message , but it is impossible to guess the message from the hash-string
        - We cannot reverse engineer the message from the hash-string
        - Even if we could the possibility of finding the string should be astronomical
    - Deterministic :
        - It must provide same output for same input everytime
    - Fast Computation :
        - The computation must be fast
        - beacuse Blockchain cannot withstand with slow hashing function
    - Withstand Collisions : 
        - It must not provide same hash-string for two differernt inputs
    - Avlanche effect : 
        - A small change in input message should trigger unpredictable and huge change in resulting hash-string

- Some examples of cryptogarphic hash functions are : 
    - SHA-1
    - SHA-2
    - SHA-3
    - SHA-256
    - Keccak-256
    - MD5
    - BLAKE-2
    - BLAKE-3
    - Whirlpool

- Blockchains are held together by the hash of previous block



- The first block in a blockchain, which has no preceding block is known as `Genesis Block`
- hash string of a block is like a fingerprint of that block



- SHA256 produces 64 hexadecimal charecters
- Since each hexadecimal charecter is a 4 bit charecter, hence SHA256 produces an output of 256 bits
