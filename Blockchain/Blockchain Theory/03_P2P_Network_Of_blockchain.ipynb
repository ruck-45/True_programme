# Immutable ledger

- Blockchain is said to be immutable ledger
- Immutable means it cannot be cahanged 
- This makes it a lot resistant to hacking

- It is immutable because each block is dependent on the hash of the previous block
- If a hacker changes a block information , then the hash of that block changes
- The change in hash of one block affects subsequent blocks following the block
- This makes the chain invalid from the point of hacking

# Centralized Vs P2P (peer to peer) network

### Centralized Network
- In centralized network all information is stored in a single system known as Server
- Every other system accessing the information stored in the server are Clients
- To get information, clients send a request signal to the server
- Upon receiving the request, server sends the file requested in the response signal

- Some advantages of centralized network are : 
    - They are easy to maintain
    - Avoid expensive duplication of work
    - Consistent
    - Cheap
    - Easy administration
    - Efficent

- Disadvantages Of centralized network : 
    - Autonmous authority which can lead to monopoly
    - Easy to hack (only one system isneeded to be hacked)
    - Insecure and requirement of backups

### Distributed / P2P / Decentralized Network
- P2P is a decentralized form of network
- Here no on is server and client
- Instead all are peers
- Everyone acts as a self sufficient server, having all data themselves
- Whenever someone needs information all they have to do is to join the network as a peer and get all information from his peers
- Peers interact with each other for information sharing

# P2P in Blockchain

- Blockchain is a decentralized network
- Means everyone has a copy of the blockchain of their own
- When one adds a new block to his ledger / blockchain, he broadcasts it to hsi peers
- First they validate if the block is ligit or fake, then they add it to their chain and continue to broadcast

# Hacking a decentralized network

- The reason of decentralized network to be secure is beacuse everyone has the copy of the data
- In centralized system a hacker could hack the server and change the data for everyone
- But to hack a decentralized system, a hacker has to hack all computers simultaneously
- And the more number of peers the network has the more difficult it will be for the hacker to hack the network

- Suppose a hacker managed to hack one of the system
- Since the hacker is smart, he also managed to hack furthur block and make them valid
- So the person's system couldnot recognize that the chain it contains is hacked
- But its peers could recongize, as the hash of end block will not match
- Since the odds of the chain of the given system is hacked is against its peers, the hacked chain will get replaced due to majority
- Hence blockchain will revert to its original form
